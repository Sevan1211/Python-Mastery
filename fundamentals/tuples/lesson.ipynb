## 1. What Is a Tuple?

A tuple is an **ordered, immutable, heterogeneous** sequence.

- **Ordered**: Items have positions (indices). Order is preserved.
- **Immutable**: You **cannot** add, remove, or change elements after creation.
- **Heterogeneous**: A tuple can hold different types.

Think of a tuple as a **frozen list** — once created, it's locked.

```python
# A tuple is defined with parentheses (or just commas)
point = (3, 7)
print(type(point))  # <class 'tuple'>
print(point)        # (3, 7)
print(len(point))   # 2
```

## 2. Creating Tuples

There are several ways to create tuples. The tricky part is single-element tuples.

```python
# --- Parentheses (most common) ---
colors = ("red", "green", "blue")
print("Parens:", colors)

# --- Without parentheses (comma-separated) ---
# The COMMAS make it a tuple, not the parentheses!
coords = 10, 20, 30
print("No parens:", coords)
print(type(coords))  # <class 'tuple'>

# --- Empty tuple ---
empty1 = ()
empty2 = tuple()
print("Empty:", empty1, empty2)

# --- From a list ---
from_list = tuple([1, 2, 3])
print("From list:", from_list)  # (1, 2, 3)

# --- From a string ---
from_str = tuple("hello")
print("From string:", from_str)  # ('h', 'e', 'l', 'l', 'o')

# --- From range ---
from_range = tuple(range(5))
print("From range:", from_range)  # (0, 1, 2, 3, 4)

# --- Mixed types ---
mixed = (1, "hello", 3.14, True, None)
print("Mixed:", mixed)
```

---
## 3. The Single-Element Trap

This is the **#1 tuple gotcha**. A single value in parentheses is NOT a tuple — you need a trailing comma.

```python
# ❌ NOT a tuple — just an int in parentheses
not_a_tuple = (5)
print(type(not_a_tuple))  # <class 'int'>
print(not_a_tuple)        # 5

# ✅ Single-element tuple — trailing comma required!
single = (5,)
print(type(single))  # <class 'tuple'>
print(single)        # (5,)

# Also works without parens
single2 = 5,
print(type(single2))  # <class 'tuple'>

# This matters in practice:
print("hello" * (3))   # hellohellohello (int 3)
print("hello" * (3,))  # TypeError! Can't multiply by tuple
```

## 4. Indexing & Slicing

Identical to lists and strings. Zero-based, negative indexing, `[start:stop:step]`.

```python
data = (10, 20, 30, 40, 50)

# Indexing
print(data[0])     # 10
print(data[-1])    # 50
print(data[2])     # 30

# Slicing — returns a NEW TUPLE (not a list!)
print(data[1:4])   # (20, 30, 40)
print(data[:3])    # (10, 20, 30)
print(data[::2])   # (10, 30, 50)
print(data[::-1])  # (50, 40, 30, 20, 10)

# Out-of-bounds slicing is safe
print(data[100:200])  # ()

# Out-of-bounds indexing raises error
try:
    print(data[10])
except IndexError as e:
    print(f"IndexError: {e}")
```

## 5. Immutability — The Core Difference

This is **the** defining feature of tuples. Once created, you cannot modify them.

```python
t = (1, 2, 3)

# ❌ Cannot change elements
try:
    t[0] = 99
except TypeError as e:
    print(f"TypeError: {e}")  # 'tuple' object does not support item assignment

# ❌ Cannot append, extend, insert, remove, pop, sort, reverse
# None of these methods exist on tuples
print(dir(t))  # Very few methods compared to lists
```

```python
# IMPORTANT NUANCE: Immutability is SHALLOW
# A tuple can contain mutable objects, and those CAN be changed!

t = (1, 2, [3, 4])
print(t)  # (1, 2, [3, 4])

# Can't replace the list with something else
try:
    t[2] = [5, 6]
except TypeError as e:
    print(f"Can't replace: {e}")

# BUT — can mutate the list INSIDE the tuple!
t[2].append(5)
print(t)  # (1, 2, [3, 4, 5])  — the list inside changed!

# The tuple's reference to the list didn't change.
# The list itself was mutated. This is a subtle but critical distinction.
```

```python
# To "modify" a tuple, create a new one
original = (1, 2, 3)

# "Add" an element → concatenate
new = original + (4,)
print(new)  # (1, 2, 3, 4)

# "Replace" element at index 1 → rebuild
modified = original[:1] + (99,) + original[2:]
print(modified)  # (1, 99, 3)

# "Remove" element at index 1 → rebuild
removed = original[:1] + original[2:]
print(removed)  # (1, 3)

# Original is always unchanged
print(original)  # (1, 2, 3)
```

## 6. Tuple Operators

| Operator | Meaning | Example |
|----------|---------|--------|
| `+` | Concatenation | `(1,2) + (3,4)` → `(1,2,3,4)` |
| `*` | Repetition | `(0,) * 3` → `(0,0,0)` |
| `in` | Membership | `3 in (1,2,3)` → `True` |
| `==` | Equality | `(1,2) == (1,2)` → `True` |
| `<` `>` | Comparison | `(1,2) < (1,3)` → `True` |

```python
# Concatenation — creates a NEW tuple
a = (1, 2)
b = (3, 4)
c = a + b
print(c)  # (1, 2, 3, 4)

# Repetition
print(("ha",) * 3)  # ('ha', 'ha', 'ha')
# Don't forget the comma! ("ha") * 3 → "hahaha" (string!)

# Membership
print(3 in (1, 2, 3))       # True
print("x" not in ("a", "b")) # True

# Comparison — element by element (lexicographic)
print((1, 2, 3) < (1, 2, 4))   # True (3 < 4)
print((1, 2) < (1, 2, 0))      # True (shorter is "less")
print((1, 2, 3) == (1, 2, 3))  # True
```

## 7. Tuple Methods (Only 2!)

Since tuples are immutable, they only have 2 methods. Compare this to lists which have 11+.

```python
t = (1, 3, 5, 3, 7, 3, 9)

# --- count(x) — how many times x appears ---
print(t.count(3))   # 3
print(t.count(99))  # 0

# --- index(x) — index of first occurrence ---
print(t.index(3))     # 1 (first 3 is at index 1)
print(t.index(3, 2))  # 3 (search starting from index 2)

try:
    t.index(99)  # ValueError if not found
except ValueError as e:
    print(f"ValueError: {e}")

# That's it! No append, insert, extend, remove, pop, sort, reverse, clear, copy
print(f"\nList methods: {len([m for m in dir([]) if not m.startswith('_')])}")
print(f"Tuple methods: {len([m for m in dir(()) if not m.startswith('_')])}")
```

## 8. Packing & Unpacking

This is where tuples truly shine. Packing and unpacking are used **everywhere** in Python — function returns, loop iteration, variable swaps.

```python
# --- Packing: multiple values → tuple ---
packed = 1, 2, 3  # No parentheses needed
print(packed)       # (1, 2, 3)
print(type(packed)) # <class 'tuple'>
```

```python
# --- Unpacking: tuple → multiple variables ---
x, y, z = (10, 20, 30)
print(x, y, z)  # 10 20 30

# Works without parens too
a, b, c = 1, 2, 3
print(a, b, c)  # 1 2 3

# Must match the count!
try:
    a, b = (1, 2, 3)  # Too many values
except ValueError as e:
    print(f"ValueError: {e}")
```

```python
# --- Star unpacking: capture the rest ---
first, *rest = (1, 2, 3, 4, 5)
print(first)  # 1
print(rest)   # [2, 3, 4, 5]  ← NOTE: rest is a LIST, not a tuple!

*beginning, last = (1, 2, 3, 4, 5)
print(beginning)  # [1, 2, 3, 4]
print(last)       # 5

first, *middle, last = (1, 2, 3, 4, 5)
print(first)   # 1
print(middle)  # [2, 3, 4]  ← list!
print(last)    # 5
```

```python
# --- Unpacking in loops — VERY common ---
students = [
    ("Alice", 90),
    ("Bob", 85),
    ("Charlie", 92)
]

for name, score in students:
    print(f"{name}: {score}")

# This works because each element is a tuple that gets unpacked
```

```python
# --- Ignoring values with _ ---
data = ("Alice", 30, "Engineer", "NYC")

name, _, job, _ = data  # _ means "don't care"
print(name, job)  # Alice Engineer

# Ignore multiple values
name, *_ = data
print(name)  # Alice
```

## 9. Swapping with Tuples

Python's elegant variable swap uses tuple packing/unpacking under the hood.

```python
# Traditional swap (other languages)
a, b = 1, 2
temp = a
a = b
b = temp
print(a, b)  # 2 1

# Pythonic swap — tuple pack/unpack
a, b = 1, 2
a, b = b, a  # Right side packs into (2, 1), then unpacks
print(a, b)  # 2 1

# Works with 3+ variables too
a, b, c = 1, 2, 3
a, b, c = c, a, b  # Rotate
print(a, b, c)  # 3 1 2
```

## 10. Tuples as Dictionary Keys

Because tuples are **immutable** and **hashable**, they can be used as dictionary keys. Lists cannot.

```python
# Tuple as dict key — works!
locations = {
    (40.7128, -74.0060): "New York",
    (34.0522, -118.2437): "Los Angeles",
    (41.8781, -87.6298): "Chicago"
}

print(locations[(40.7128, -74.0060)])  # New York

# List as dict key — FAILS!
try:
    bad = {[1, 2]: "nope"}
except TypeError as e:
    print(f"TypeError: {e}")  # unhashable type: 'list'
```

```python
# Common use: counting pairs or coordinates
from collections import Counter

moves = [(0, 1), (1, 0), (0, 1), (0, -1), (1, 0), (0, 1)]
move_counts = Counter(moves)
print(move_counts)
# Counter({(0, 1): 3, (1, 0): 2, (0, -1): 1})

# Also works in sets
visited = set()
visited.add((0, 0))
visited.add((1, 0))
visited.add((0, 0))  # duplicate — ignored
print(visited)  # {(0, 0), (1, 0)}
```

```python
# GOTCHA: Tuple with mutable content is NOT hashable
try:
    bad_key = {(1, [2, 3]): "value"}  # list inside tuple → unhashable
except TypeError as e:
    print(f"TypeError: {e}")

# Only tuples containing ALL immutable elements are hashable
```

## 11. Tuples as Return Values

Functions often return multiple values as tuples. This is one of the most common uses.

```python
# Returning multiple values
def min_max(nums):
    return min(nums), max(nums)  # Returns a tuple

result = min_max([3, 1, 4, 1, 5, 9])
print(result)        # (1, 9)
print(type(result))  # <class 'tuple'>

# Unpack directly
lo, hi = min_max([3, 1, 4, 1, 5, 9])
print(f"min={lo}, max={hi}")  # min=1, max=9
```

```python
# Common pattern: return status + value
def safe_divide(a, b):
    if b == 0:
        return False, None
    return True, a / b

success, result = safe_divide(10, 3)
print(f"Success: {success}, Result: {result:.2f}")

success, result = safe_divide(10, 0)
print(f"Success: {success}, Result: {result}")
```

```python
# divmod() — built-in that returns a tuple
quotient, remainder = divmod(17, 5)
print(f"17 ÷ 5 = {quotient} remainder {remainder}")  # 3 remainder 2

# enumerate() returns tuples
for item in enumerate(["a", "b", "c"]):
    print(item, type(item))  # (0, 'a') <class 'tuple'>  etc.
```

## 12. Named Tuples

Named tuples give your tuple fields **descriptive names** instead of numeric indices. They're like lightweight classes.

```python
from collections import namedtuple

# Define a named tuple type
Point = namedtuple("Point", ["x", "y"])

# Create instances
p = Point(3, 7)
print(p)        # Point(x=3, y=7)
print(p.x)      # 3  — access by name
print(p.y)      # 7
print(p[0])     # 3  — still works by index

# Still immutable
try:
    p.x = 99
except AttributeError as e:
    print(f"AttributeError: {e}")

# Still supports unpacking
x, y = p
print(f"x={x}, y={y}")
```

```python
# Practical example: representing structured records
Student = namedtuple("Student", ["name", "age", "grade"])

students = [
    Student("Alice", 20, "A"),
    Student("Bob", 22, "B"),
    Student("Charlie", 21, "A"),
]

# Much more readable than students[i][0]
for s in students:
    print(f"{s.name} (age {s.age}): grade {s.grade}")

# Can use as dict key (still hashable)
print(f"\nHash of Alice: {hash(students[0])}")
```

```python
# _replace() — create a modified copy (since tuples are immutable)
p = Point(3, 7)
p2 = p._replace(x=10)
print(p)   # Point(x=3, y=7)  — unchanged
print(p2)  # Point(x=10, y=7) — new copy

# _asdict() — convert to dict
print(p._asdict())  # {'x': 3, 'y': 7}

# _fields — get field names
print(Point._fields)  # ('x', 'y')
```

## 13. Tuples vs Lists — When to Use Which

| | Tuple | List |
|---|---|---|
| Mutable? | ❌ No | ✅ Yes |
| Hashable? | ✅ Yes (if all elements immutable) | ❌ No |
| Dict key? | ✅ Yes | ❌ No |
| Set element? | ✅ Yes | ❌ No |
| Methods | 2 (count, index) | 11+ |
| Memory | Less | More |
| Speed | Slightly faster | Slightly slower |
| Use for | Fixed collections, records, keys | Dynamic collections |

### Rules of Thumb

- **Use a tuple** when data shouldn't change: coordinates, RGB colors, database rows, function returns
- **Use a list** when you'll add, remove, or modify elements: shopping lists, queue of tasks, building results
- **Tuple = struct** (different fields with meaning), **List = array** (same type, variable length)

```python
import sys

# Memory comparison
lst = [1, 2, 3, 4, 5]
tup = (1, 2, 3, 4, 5)

print(f"List size: {sys.getsizeof(lst)} bytes")
print(f"Tuple size: {sys.getsizeof(tup)} bytes")
# Tuple is smaller! Lists need extra space for dynamic resizing.
```

## 14. Common Built-ins with Tuples

All the same built-ins that work with lists work with tuples.

```python
nums = (4, 7, 2, 9, 1, 5, 8)

print(f"len:  {len(nums)}")    # 7
print(f"min:  {min(nums)}")    # 1
print(f"max:  {max(nums)}")    # 9
print(f"sum:  {sum(nums)}")    # 36
print(f"sorted: {sorted(nums)}")  # [1, 2, 4, 5, 7, 8, 9]  ← returns a LIST

# any / all
print(f"any: {any(nums)}")   # True
print(f"all: {all(nums)}")   # True (no zeros)

# enumerate works with tuples
for i, val in enumerate(nums):
    print(f"  {i}: {val}")

# zip works with tuples
names = ("Alice", "Bob")
scores = (90, 85)
for name, score in zip(names, scores):
    print(f"  {name}: {score}")
```

## 15. Traversal Patterns

All traversal patterns from lists work identically with tuples. Here are the most common ones.

```python
data = (10, 20, 30, 40, 50)

# Forward by value
for val in data:
    print(val, end=" ")
print()

# With index
for i, val in enumerate(data):
    print(f"{i}:{val}", end=" ")
print()

# Reverse
for val in reversed(data):
    print(val, end=" ")
print()
```

```python
# Unpacking in iteration — the tuple superpower
records = (
    ("Alice", "Engineering", 95000),
    ("Bob", "Marketing", 72000),
    ("Charlie", "Engineering", 88000),
)

# Pattern: iterate and unpack simultaneously
for name, dept, salary in records:
    print(f"{name:10s} | {dept:15s} | ${salary:,}")

# Filter pattern: build a new tuple from existing
engineers = tuple(name for name, dept, _ in records if dept == "Engineering")
print(f"\nEngineers: {engineers}")
```

```python
# Sorting tuples — lexicographic by default
students = [
    ("Charlie", 92),
    ("Alice", 85),
    ("Bob", 92),
    ("Alice", 78),
]

# Sorts by first element, then second (like alphabetical then numeric)
print(sorted(students))
# [('Alice', 78), ('Alice', 85), ('Bob', 92), ('Charlie', 92)]

# Sort by score (second element)
print(sorted(students, key=lambda s: s[1]))
# [('Alice', 78), ('Alice', 85), ('Bob', 92), ('Charlie', 92)]

# Sort by score descending, then name ascending
print(sorted(students, key=lambda s: (-s[1], s[0])))
# [('Bob', 92), ('Charlie', 92), ('Alice', 85), ('Alice', 78)]
```

## 16. Gotchas & Pitfalls

```python
# Gotcha 1: Single-element tuple needs trailing comma
not_tuple = (42)
is_tuple = (42,)
print(type(not_tuple))  # <class 'int'>
print(type(is_tuple))   # <class 'tuple'>
```

```python
# Gotcha 2: Accidental tuple creation with trailing comma
x = 42,  # This is a tuple! Not an int!
print(type(x))  # <class 'tuple'>
print(x)        # (42,)

# This can cause subtle bugs in math operations
try:
    result = x + 1  # TypeError!
except TypeError as e:
    print(f"TypeError: {e}")
```

```python
# Gotcha 3: sorted() returns a LIST, not a tuple
t = (5, 3, 1, 4, 2)
result = sorted(t)
print(type(result))  # <class 'list'> — NOT tuple!
print(result)        # [1, 2, 3, 4, 5]

# To get a sorted tuple:
sorted_tuple = tuple(sorted(t))
print(type(sorted_tuple))  # <class 'tuple'>
print(sorted_tuple)        # (1, 2, 3, 4, 5)
```

```python
# Gotcha 4: Star unpacking always produces a LIST
first, *rest = (1, 2, 3, 4)
print(type(rest))  # <class 'list'> — not tuple!
print(rest)        # [2, 3, 4]
```

```python
# Gotcha 5: Mutable objects inside tuples
t = ([1, 2], [3, 4])

# Can't reassign
try:
    t[0] = [5, 6]
except TypeError:
    print("Can't reassign")

# CAN mutate the inner list
t[0].append(99)
print(t)  # ([1, 2, 99], [3, 4])

# += on inner list is WEIRD:
try:
    t[0] += [100]  # This BOTH mutates AND raises TypeError!
except TypeError:
    print(f"Error raised, but t is now: {t}")
    # t = ([1, 2, 99, 100], [3, 4]) — it DID mutate!
```

## 17. Real-World Use Cases

```python
# --- Coordinate systems ---
origin = (0, 0)
destination = (3, 4)

# Distance calculation
distance = ((destination[0] - origin[0])**2 + (destination[1] - origin[1])**2) ** 0.5
print(f"Distance: {distance}")  # 5.0

# --- RGB colors ---
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)
WHITE = (255, 255, 255)

# --- Database rows (each row is a tuple) ---
# This is how SQLite and many ORMs return data
rows = [
    (1, "Alice", "alice@email.com", True),
    (2, "Bob", "bob@email.com", False),
    (3, "Charlie", "charlie@email.com", True),
]

for user_id, name, email, active in rows:
    status = "active" if active else "inactive"
    print(f"  [{user_id}] {name} ({email}) — {status}")
```

```python
# --- Config / constants ---
ALLOWED_EXTENSIONS = (".py", ".ipynb", ".md", ".txt")
HTTP_SUCCESS_CODES = (200, 201, 204)

filename = "lesson.ipynb"
if filename.endswith(ALLOWED_EXTENSIONS):  # endswith accepts tuple!
    print(f"{filename} is allowed")

# --- Interview pattern: tracking visited states ---
# e.g., BFS on a grid where state = (row, col)
visited = set()
visited.add((0, 0))
visited.add((0, 1))
print((0, 0) in visited)  # True — O(1) lookup

# --- Sorting with multiple criteria ---
tasks = [
    ("High", 1, "Fix bug"),
    ("Low", 3, "Update docs"),
    ("High", 2, "Add feature"),
]
# Sort by priority then by number
priority_order = {"High": 0, "Medium": 1, "Low": 2}
sorted_tasks = sorted(tasks, key=lambda t: (priority_order[t[0]], t[1]))
for priority, num, desc in sorted_tasks:
    print(f"  [{priority}] #{num}: {desc}")
```

## 18. Summary

### Key Takeaways

1. **Tuples are immutable lists** — once created, they cannot be changed.
2. **Commas make tuples, not parentheses** — `(5)` is an int, `(5,)` is a tuple.
3. **Only 2 methods**: `count()` and `index()`.
4. **Hashable** (if all elements are immutable) → can be dict keys and set elements.
5. **Packing/unpacking is the killer feature** — used everywhere in Python.
6. **Functions return tuples** for multiple return values.
7. **Named tuples** give fields names for readability.
8. **Immutability is shallow** — mutable objects inside a tuple can still be changed.
9. **sorted() returns a list**, not a tuple. Wrap in `tuple()` if needed.
10. **Star unpacking produces a list**, not a tuple.

### When to Use Tuples

- Fixed collections that shouldn't change (coordinates, colors, config)
- Dictionary keys or set elements
- Function return values
- Records/rows from databases
- Anywhere you want to signal "this data is read-only"
