```python
# Problem 1: Reverse a String
# Return the reverse of the input string. Do NOT use slicing [::-1].

def reverse_string(s):
    return s[::-1]

assert reverse_string("hello") == "olleh"
assert reverse_string("") == ""
assert reverse_string("a") == "a"
assert reverse_string("ab") == "ba"
assert reverse_string("racecar") == "racecar"
assert reverse_string("12345") == "54321"
assert reverse_string("  spaces  ") == "  secaps  "
print("Problem 1 passed ✓")
```

```python
# Problem 2: Count Vowels
# Return the number of vowels (a, e, i, o, u) in a string. Case-insensitive.

def count_vowels(s):
    vowels = 'aeiou'
    count = 0

    for ch in s:
        if ch.lower() in set(vowels):
            count += 1
    return count


assert count_vowels("hello") == 2
assert count_vowels("AEIOU") == 5
assert count_vowels("") == 0
assert count_vowels("bcdfg") == 0
assert count_vowels("aEiOu") == 5
assert count_vowels("Python") == 1
assert count_vowels("aaa") == 3
print("Problem 2 passed ✓")
```

```python
# Problem 4: Sum of Even Numbers
# Return the sum of all even numbers in a list.

def sum_evens(nums):
    total = 0

    for num in nums:
        if num % 2 == 0:
            total += num
    return total

assert sum_evens([1, 2, 3, 4]) == 6
assert sum_evens([]) == 0
assert sum_evens([1, 3, 5]) == 0
assert sum_evens([2, 4, 6]) == 12
assert sum_evens([0]) == 0
assert sum_evens([-2, -4, 1]) == -6
assert sum_evens([10, 21, 32, 43, 54]) == 96
print("Problem 4 passed ✓")
```

```python
# Problem 5: Check Palindrome
# Return True if the string is a palindrome (case-insensitive, letters only).

def is_palindrome(s):
    return s.lower().split() == s[::-1].lower().split()

assert is_palindrome("racecar") == True
assert is_palindrome("hello") == False
assert is_palindrome("") == True
assert is_palindrome("a") == True
assert is_palindrome("Aba") == True
assert is_palindrome("A man a plan a canal Panama") == True
assert is_palindrome("not a palindrome") == False
print("Problem 5 passed ✓")
```

```python
# Problem 6: Remove Duplicates from List
# Return a new list with duplicates removed, preserving original order.

def remove_duplicates(lst):
    seen = set()
    result = []

    for i in lst:
        if i in seen:
            continue
        seen.add(i)
        result.append(i)

    return result 


assert remove_duplicates([1, 2, 2, 3]) == [1, 2, 3]
assert remove_duplicates([]) == []
assert remove_duplicates([1, 1, 1]) == [1]
assert remove_duplicates([1, 2, 3]) == [1, 2, 3]
assert remove_duplicates([3, 1, 2, 1, 3]) == [3, 1, 2]
assert remove_duplicates(["a", "b", "a"]) == ["a", "b"]
assert remove_duplicates([5]) == [5]
print("Problem 6 passed ✓")
```

```python
# Problem 7: FizzBuzz List
# Return a list from 1 to n where:
# - multiples of 3 are "Fizz"
# - multiples of 5 are "Buzz"
# - multiples of both are "FizzBuzz"
# - others are the number as a string

def fizzbuzz(n):
    result = []
    for i in range(1, n+1):
        if i % 3 == 0 and i % 5 ==0:
            result.append("FizzBuzz")
        elif i % 3 == 0:
            result.append("Fizz")
        elif i % 5 == 0:
            result.append("Buzz")
        else:
            result.append(str(i))
    return result


assert fizzbuzz(1) == ["1"]
assert fizzbuzz(3) == ["1", "2", "Fizz"]
assert fizzbuzz(5) == ["1", "2", "Fizz", "4", "Buzz"]
assert fizzbuzz(15)[-1] == "FizzBuzz"
assert fizzbuzz(15)[2] == "Fizz"
assert fizzbuzz(15)[4] == "Buzz"
assert len(fizzbuzz(100)) == 100
print("Problem 7 passed ✓")
```

```python
# Problem 9: Flatten a List of Lists
# Given a list of lists, return a single flat list.

def flatten(lst):
    result = []

    for sublist in lst:
        for item in sublist:
            result.append(item)
    return result

assert flatten([[1, 2], [3, 4]]) == [1, 2, 3, 4]
assert flatten([]) == []
assert flatten([[], []]) == []
assert flatten([[1]]) == [1]
assert flatten([[1, 2, 3], [4], [5, 6]]) == [1, 2, 3, 4, 5, 6]
assert flatten([["a", "b"], ["c"]]) == ["a", "b", "c"]
assert flatten([[], [1], [], [2, 3], []]) == [1, 2, 3]
print("Problem 9 passed ✓")
```

```python
# Problem 11: Title Case
# Capitalize the first letter of each word in a string. Do NOT use .title().

def title_case(s):
    words = s.split()
    result = []

    for word in words:
        result.append(word[0].upper() +  word[1:])

    return " ".join(result)


assert title_case("hello world") == "Hello World"
assert title_case("") == ""
assert title_case("a") == "A"
assert title_case("HELLO") == "HELLO"
assert title_case("hello") == "Hello"
assert title_case("one two three") == "One Two Three"
assert title_case("already Good") == "Already Good"
print("Problem 11 passed ✓")
```

```python
# Problem 15: Rotate List
# Rotate a list to the right by k positions.

def rotate_list(lst, k):
    if not lst:
        return []

    k = k % len(lst)
    
    last = lst[-k:]
    first = lst[:-k]

    return list(last + first)

assert rotate_list([1, 2, 3, 4, 5], 2) == [4, 5, 1, 2, 3]
assert rotate_list([1, 2, 3], 0) == [1, 2, 3]
assert rotate_list([], 5) == []
assert rotate_list([1], 10) == [1]
assert rotate_list([1, 2, 3], 3) == [1, 2, 3]
assert rotate_list([1, 2, 3, 4], 1) == [4, 1, 2, 3]
assert rotate_list([1, 2], 5) == [2, 1]
print("Problem 15 passed ✓")
```

```python
# Problem 17: Check if List is Sorted
# Return True if the list is sorted in non-decreasing order.

def is_sorted(nums):
    for i, num in enumerate(nums):
        if i != 0 and num < nums[i-1]:
            return False
        
    return True

assert is_sorted([1, 2, 3]) == True
assert is_sorted([3, 2, 1]) == False
assert is_sorted([]) == True
assert is_sorted([1]) == True
assert is_sorted([1, 1, 1]) == True
assert is_sorted([1, 3, 2]) == False
assert is_sorted([-5, -3, 0, 4]) == True
print("Problem 17 passed ✓")
```

```python
# Problem 18: Zip Two Lists into a Dict
# Given two lists of equal length, return a dict pairing them.

def zip_to_dict(keys, values):
    result = {}

    for k, v in zip(keys, values):
        result[k] = v
    return result

assert zip_to_dict(["a", "b"], [1, 2]) == {"a": 1, "b": 2}
assert zip_to_dict([], []) == {}
assert zip_to_dict([1], ["x"]) == {1: "x"}
assert zip_to_dict(["x", "y", "z"], [10, 20, 30]) == {"x": 10, "y": 20, "z": 30}
assert zip_to_dict(["a", "a"], [1, 2]) == {"a": 2}
assert zip_to_dict([True, False], [1, 0]) == {True: 1, False: 0}
assert zip_to_dict(["name"], ["Alice"]) == {"name": "Alice"}
print("Problem 18 passed ✓")
```
