```python
# Problem 29: Implement a Retry Decorator
# A decorator that retries a function up to `max_retries` times on exception.
# If all retries fail, raise the last exception.

def retry(max_retries=3):
    pass  # return decorator

attempt_count = 0

@retry(max_retries=3)
def flaky_function():
    global attempt_count
    attempt_count += 1
    if attempt_count < 3:
        raise ValueError("not yet")
    return "success"

attempt_count = 0
assert flaky_function() == "success"
assert attempt_count == 3

fail_count = 0
@retry(max_retries=2)
def always_fails():
    global fail_count
    fail_count += 1
    raise RuntimeError("fail")

fail_count = 0
try:
    always_fails()
    assert False, "Should have raised"
except RuntimeError:
    pass
assert fail_count == 2

@retry(max_retries=1)
def immediate_success():
    return 42
assert immediate_success() == 42
print("Problem 29 passed ✓")
```
