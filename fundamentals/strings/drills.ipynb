## Section 5: Build from Scratch

```python
# BUILD: Write a function that counts vowels (a, e, i, o, u) in a string.
# Should be case-insensitive.

def count_vowels(s):
    # TODO
    vowels = "aeiou"
    count = 0

    for ch in s:
        if ch.lower() in vowels:
            count += 1
    return count 


assert count_vowels("hello") == 2
assert count_vowels("AEIOU") == 5
assert count_vowels("rhythm") == 0
assert count_vowels("") == 0
assert count_vowels("aEiOu") == 5
print("count_vowels passed ✓")
```

```python
# BUILD: Write a function that reverses the words in a string.
# "hello world foo" → "foo world hello"
# Multiple spaces should become single spaces.

def reverse_words(s):
    # TODO
    words = s.split()
    words.reverse()
    return ' '.join(words)

assert reverse_words("hello world foo") == "foo world hello"
assert reverse_words("  hello   world  ") == "world hello"
assert reverse_words("single") == "single"
assert reverse_words("") == ""
print("reverse_words passed ✓")
```

```python
# BUILD: Write a function that checks if a string is a palindrome.
# Ignore case and non-alphanumeric characters.
# "A man, a plan, a canal: Panama" → True

def is_palindrome(s):
    clean = "".join(char.lower() for char in s if char.isalnum())
    return clean == clean[::-1]

assert is_palindrome("racecar") == True
assert is_palindrome("A man, a plan, a canal: Panama") == True
assert is_palindrome("hello") == False
assert is_palindrome("") == True
assert is_palindrome("a") == True
assert is_palindrome("Was it a car or a cat I saw?") == True
print("is_palindrome passed ✓")
```

```python
# BUILD: Write a function that returns the first non-repeating character.
# "aabbcdd" → "c"
# Return "" if all characters repeat.

def first_unique_char(s):
    # TODO
    counts = {}

    for ch in s:
        counts[ch] = counts.get(ch, 0) + 1

    for ch in s:
        if counts[ch] == 1:
            return ch

    return ""

assert first_unique_char("aabbcdd") == "c"
assert first_unique_char("abcabc") == ""
assert first_unique_char("leetcode") == "l"
assert first_unique_char("") == ""
assert first_unique_char("z") == "z"
print("first_unique_char passed ✓")
```

## Section 6 — Explore & Visualize

Run each cell and study the output. These cells help you build intuition for how strings
work under the hood.

```python
# EXPLORE: String immutability — every operation creates a NEW string.
# Watch the id() change with each operation.

s = "hello"
print(f"Original:  id={id(s)}  value='{s}'")

s_upper = s.upper()
print(f"After .upper(): id={id(s_upper)}  value='{s_upper}'")
print(f"Original unchanged: id={id(s)}  value='{s}'")

s2 = s + " world"
print(f"After concat: id={id(s2)}  value='{s2}'")
print(f"Original unchanged: id={id(s)}  value='{s}'")

# KEY INSIGHT: s never changes. Every method returns a NEW string object.
# This is why you must reassign: s = s.upper(), not just s.upper()
```

```python
# EXPLORE: ASCII values and character ordering
# ord() converts char → int, chr() converts int → char

print("=== Lowercase letters ===")
for c in "abcdefghijklmnopqrstuvwxyz":
    print(f"  '{c}' = {ord(c)}", end="")
print()

print("\n=== Uppercase letters ===")
for c in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
    print(f"  '{c}' = {ord(c)}", end="")
print()

print("\n=== Digits ===")
for c in "0123456789":
    print(f"  '{c}' = {ord(c)}", end="")
print()

print(f"\n'A' < 'a' → {'A' < 'a'}  (because {ord('A')} < {ord('a')})")
print(f"'Z' < 'a' → {'Z' < 'a'}  (because {ord('Z')} < {ord('a')})")
print(f"'9' < 'A' → {'9' < 'A'}  (because {ord('9')} < {ord('A')})")

# KEY INSIGHT: Python compares strings character-by-character using Unicode/ASCII values.
# Digits < Uppercase < Lowercase
```

```python
# EXPLORE: String concatenation performance
# Building a string with += in a loop is SLOW for large strings.
# .join() is FAST because it pre-allocates memory.

import time

n = 50_000

# Slow way: += in a loop
start = time.perf_counter()
result = ""
for i in range(n):
    result += "a"
slow_time = time.perf_counter() - start

# Fast way: join
start = time.perf_counter()
result = "".join("a" for i in range(n))
fast_time = time.perf_counter() - start

print(f"+= loop:  {slow_time:.4f}s")
print(f".join():  {fast_time:.4f}s")
print(f"join() is ~{slow_time / fast_time:.1f}x faster")

# KEY INSIGHT: In interviews, if you're building a string in a loop,
# collect pieces in a list and .join() at the end.
```
