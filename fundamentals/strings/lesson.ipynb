## 1. Intuition — What Is a String?

A string is a **sequence of characters**. Think of it as an ordered array where each element is a single character.

Key properties:
- **Immutable** — once created, you cannot change individual characters
- **Iterable** — you can loop over them character by character
- **Indexable** — you can access any character by position
- **Sliceable** — you can extract sub-sequences

```
String:   P  y  t  h  o  n
Index:    0  1  2  3  4  5
Negative: -6 -5 -4 -3 -2 -1
```

````python
# A string is a sequence of characters
name = "Python"

print(f"String: {name}")
print(f"Type: {type(name)}")
print(f"Length: {len(name)}")
print(f"Is it a sequence? Let's see each character:")
print(list(name))  # Convert to list to see individual characters
````

## 2. String Creation

Python gives you several ways to create strings. Each has its use case.

````python
# Single quotes and double quotes are identical
s1 = 'hello'
s2 = "hello"
print(f"Single quotes: {s1}")
print(f"Double quotes: {s2}")
print(f"Are they equal? {s1 == s2}")
````

````python
# Use the OTHER quote type to include quotes inside the string
message = "it's a beautiful day"
html = '<div class="box">hello</div>'
print(message)
print(html)
````

````python
# Triple quotes — multiline strings
paragraph = """This is
a multiline
string."""
print(paragraph)
print(f"\nLength including newlines: {len(paragraph)}")
````

````python
# Raw strings — backslashes are literal (no escape processing)
# Incredibly useful for file paths and regex patterns
path = r"C:\Users\name\documents"
regex = r"\d+\.\d+"
print(f"Raw path: {path}")
print(f"Raw regex: {regex}")

# Compare with normal string:
normal = "C:\\Users\\name\\documents"  # Need double backslashes without raw
print(f"Normal path: {normal}")
print(f"Same result? {path == normal}")
````

````python
# F-strings — inline expressions (Python 3.6+)
name = "Sevan"
age = 25
print(f"My name is {name} and I'm {age} years old")
print(f"Next year I'll be {age + 1}")
print(f"Name in caps: {name.upper()}")
print(f"Name length: {len(name)}")
````

````python
# The empty string — important edge case
empty = ""
print(f"Empty string: '{empty}'")
print(f"Length: {len(empty)}")
print(f"Bool value: {bool(empty)}")  # False — empty string is falsy
print(f"Bool of space: {bool(' ')}")  # True — space is NOT empty
print(f"Is '' equal to ' '? {'' == ' '}")
````

````python
# Adjacent string literal concatenation — Python auto-joins them
s = "hello" " " "world"
print(s)  # "hello world" — useful for breaking long strings across lines

# This is a compile-time operation, not runtime
long_url = (
    "https://api.example.com"
    "/v2/users"
    "?page=1&limit=100"
)
print(long_url)
````

## 3. Indexing

Strings are zero-indexed. You can use positive indices (from the left) or negative indices (from the right).

```
 String:   P  y  t  h  o  n
 Index:    0  1  2  3  4  5
 Negative: -6 -5 -4 -3 -2 -1
```

````python
s = "Python"

# Positive indexing
print(f"s[0] = '{s[0]}'")   # First character
print(f"s[1] = '{s[1]}'")   # Second character
print(f"s[5] = '{s[5]}'")   # Last character

# Negative indexing (counts from the end)
print(f"s[-1] = '{s[-1]}'")  # Last character
print(f"s[-2] = '{s[-2]}'")  # Second to last
print(f"s[-6] = '{s[-6]}'")  # First character
````

````python
# IndexError — accessing beyond string boundaries
s = "Python"

try:
    print(s[6])  # Only indices 0-5 exist
except IndexError as e:
    print(f"IndexError: {e}")

try:
    print(s[-7])  # Only negative indices -1 to -6 exist
except IndexError as e:
    print(f"IndexError: {e}")

# Safe way to access characters — check length first
s = "hi"
idx = 5
if idx < len(s):
    print(s[idx])
else:
    print(f"Index {idx} is out of range for string of length {len(s)}")
````

## 4. Slicing

Syntax: `string[start:stop:step]`

- `start`: inclusive (default `0`)
- `stop`: **exclusive** (default `len(s)`)
- `step`: stride (default `1`)

**Critical rule:** The stop index is always **excluded**. This is the most common source of off-by-one errors.

````python
s = "Python"

# Basic slicing
print(f"s[0:3]  = '{s[0:3]}'")   # Characters at index 0, 1, 2
print(f"s[2:5]  = '{s[2:5]}'")   # Characters at index 2, 3, 4
print(f"s[:3]   = '{s[:3]}'")    # First 3 characters (start defaults to 0)
print(f"s[3:]   = '{s[3:]}'")    # From index 3 to end
print(f"s[:]    = '{s[:]}'")     # Full copy
````

````python
s = "Python"

# Negative indices in slicing
print(f"s[-3:]   = '{s[-3:]}'")    # Last 3 characters
print(f"s[:-2]   = '{s[:-2]}'")    # Everything except last 2
print(f"s[-4:-1] = '{s[-4:-1]}'")  # From -4 to -1 (exclusive)
````

````python
s = "Python"

# Step (stride)
print(f"s[::2]   = '{s[::2]}'")    # Every other character
print(f"s[1::2]  = '{s[1::2]}'")   # Every other, starting at index 1
print(f"s[::-1]  = '{s[::-1]}'")   # REVERSED string (very common in interviews!)
print(f"s[::-2]  = '{s[::-2]}'")   # Every other, reversed
````

````python
# IMPORTANT: Slicing NEVER raises IndexError
s = "hi"
print(f"s[0:100] = '{s[0:100]}'")   # Just returns what exists
print(f"s[50:100] = '{s[50:100]}'")  # Empty string, no error
print(f"s[-100:2] = '{s[-100:2]}'")  # Clamps to start

# Compare: indexing DOES raise IndexError
try:
    s[100]
except IndexError as e:
    print(f"But s[100] raises: {e}")
````

````python
# Common slicing patterns you'll use constantly
s = "hello world"

print(f"First 5 chars:    '{s[:5]}'")      # "hello"
print(f"Last 5 chars:     '{s[-5:]}'")      # "world"
print(f"Remove first:     '{s[1:]}'")       # "ello world"
print(f"Remove last:      '{s[:-1]}'")      # "hello worl"
print(f"Reversed:         '{s[::-1]}'")     # "dlrow olleh"

# Palindrome check — one of the most common interview patterns
word = "racecar"
print(f"\n'{word}' is palindrome? {word == word[::-1]}")
````

## 5. String Immutability

This is one of the most important concepts for interviews. Strings **cannot be modified in place**. Every operation that seems to "change" a string actually creates a **brand new** string object.

````python
# You CANNOT modify a string in place
s = "hello"
try:
    s[0] = "H"  # This will fail
except TypeError as e:
    print(f"TypeError: {e}")

# You must create a new string instead
s = "H" + s[1:]
print(f"New string: '{s}'")  # "Hello"
````

````python
# Watch the id() change — every operation creates a NEW object
s = "hello"
print(f"Original:      id={id(s)}  value='{s}'")

s_upper = s.upper()
print(f"After .upper(): id={id(s_upper)}  value='{s_upper}'")
print(f"Original:      id={id(s)}  value='{s}'  (unchanged!)")

s2 = s + " world"
print(f"After concat:  id={id(s2)}  value='{s2}'")
print(f"Original:      id={id(s)}  value='{s}'  (still unchanged!)")
````

````python
# Performance implication: += in a loop is SLOW for large strings
# Because each += creates a brand new string and copies everything

import time

n = 50_000

# SLOW way: += in a loop — O(n²) total
start = time.perf_counter()
result = ""
for i in range(n):
    result += "a"
slow_time = time.perf_counter() - start

# FAST way: .join() — O(n) total
start = time.perf_counter()
result = "".join("a" for i in range(n))
fast_time = time.perf_counter() - start

print(f"+= loop:  {slow_time:.4f}s")
print(f".join():  {fast_time:.4f}s")
print(f"join() is ~{slow_time / fast_time:.1f}x faster")
print("\nRule: In interviews, always collect parts in a list and .join() at the end.")
````

## 6. String Operators

Python provides several operators that work on strings.

````python
# Concatenation (+)
greeting = "hello" + " " + "world"
print(f"Concatenation: '{greeting}'")

# You can only concatenate str with str
try:
    result = "age: " + 25
except TypeError as e:
    print(f"TypeError: {e}")

# Fix: convert to string first
print(f"Fix with str(): {'age: ' + str(25)}")
print(f"Fix with f-string: {f'age: {25}'}")
````

````python
# Repetition (*)
print("ha" * 3)       # "hahaha"
print("-" * 40)       # Line separator
print("abc" * 0)      # "" — empty string
print("abc" * -1)     # "" — negative also gives empty
````

````python
# Membership (in / not in) — used constantly in interviews
print(f"'ell' in 'hello':     {'ell' in 'hello'}")
print(f"'xyz' in 'hello':     {'xyz' in 'hello'}")
print(f"'' in 'hello':        {'' in 'hello'}")        # Empty string is in everything!
print(f"'H' in 'hello':       {'H' in 'hello'}")       # Case-sensitive!
print(f"'o' not in 'hello':   {'o' not in 'hello'}")
````

````python
# Comparison — lexicographic (character by character using Unicode values)
print(f"'apple' < 'banana':  {'apple' < 'banana'}")   # True (a < b)
print(f"'abc' < 'abd':       {'abc' < 'abd'}")         # True (c < d)
print(f"'abc' < 'abcd':      {'abc' < 'abcd'}")        # True (shorter is "less")
print(f"'Z' < 'a':           {'Z' < 'a'}")             # True (uppercase Z=90, lowercase a=97)
print(f"'10' < '9':          {'10' < '9'}")            # True! '1' < '9'

# GOTCHA: String comparison of numbers doesn't work like number comparison!
# "10" < "9" is True because Python compares '1' vs '9', and '1' comes first in ASCII.
````

## 7. String Methods — Case Manipulation

Every method returns a **new** string (strings are immutable). You must reassign if you want to keep the result.

````python
s = "hello World"

print(f"Original:    '{s}'")
print(f".upper():    '{s.upper()}'")
print(f".lower():    '{s.lower()}'")
print(f".capitalize(): '{s.capitalize()}'")  # First char upper, rest lower
print(f".title():    '{s.title()}'")
print(f".swapcase(): '{s.swapcase()}'")
print(f".casefold(): '{s.casefold()}'")
````

````python
# casefold() is more aggressive than lower() — important for international text
german = "Straße"  # German sharp s
print(f"German word: '{german}'")
print(f".lower():    '{german.lower()}'")
print(f".casefold(): '{german.casefold()}'")

# When to use which:
print("\n--- Use cases ---")
print(f".lower() for case-insensitive comparison: {'Hello'.lower() == 'hello'}")
print(f".upper() for display: {'hello'.upper()}")
print(f".capitalize() for sentences: {'hello world'.capitalize()}")
print(f".title() for names: {'john doe'.title()}")
print(f".title() gotcha: {"it's a test".title()}")  # "It'S A Test" — breaks on apostrophes
````

## 9. String Methods — Searching

These methods help you locate substrings within a string. Understanding the difference between `find()` and `index()` is important.

````python
s = "hello world hello"

# find() — returns index of first occurrence, or -1 if not found
print(f"find('world'):    {s.find('world')}")
print(f"find('xyz'):      {s.find('xyz')}")     # -1 = not found
print(f"find('hello'):    {s.find('hello')}")    # 0 = first occurrence
print(f"find('hello', 1): {s.find('hello', 1)}") # Start searching from index 1

# rfind() — search from the right
print(f"rfind('hello'):   {s.rfind('hello')}")   # 12 = last occurrence
````

````python
# index() vs find() — index() raises ValueError if not found
s = "hello world"

print(f"index('world'): {s.index('world')}")

try:
    s.index("xyz")
except ValueError as e:
    print(f"index('xyz') raises: {e}")

# Rule: Prefer find() when you're not sure the substring exists
#        Use index() when you KNOW it should be there (and want an error if not)
````

````python
# count() — count non-overlapping occurrences
s = "hello world hello"
print(f"count('hello'): {s.count('hello')}")
print(f"count('l'):     {s.count('l')}")
print(f"count('xyz'):   {s.count('xyz')}")

# GOTCHA: Non-overlapping!
print(f"'aaa'.count('aa'): {'aaa'.count('aa')}")  # 1, not 2!

# startswith() / endswith()
print(f"\nstartswith('hello'): {'hello world'.startswith('hello')}")
print(f"endswith('world'):    {'hello world'.endswith('world')}")

# Check multiple with a tuple
filename = "photo.jpg"
print(f"Is image? {filename.endswith(('.jpg', '.png', '.gif'))}")
````

## 10. String Methods — Splitting & Joining

These are among the most used string operations in both interviews and production code.

````python
# split() with no arguments — splits on ANY whitespace, strips edges
print("hello world foo".split())        # ['hello', 'world', 'foo']
print("  hello   world  ".split())      # ['hello', 'world'] — handles multiple spaces!

# split() with a delimiter
print("a,b,c".split(","))               # ['a', 'b', 'c']
print("a::b::c".split("::"))            # ['a', 'b', 'c']
print("one-two-three".split("-"))       # ['one', 'two', 'three']
````

````python
# CRITICAL DIFFERENCE: split() vs split(" ")
# This trips people up constantly!

text = "  hello   world  "
print(f"split():    {text.split()}")        # ['hello', 'world'] — clean!
print(f"split(' '): {text.split(' ')}")     # ['', '', 'hello', '', '', 'world', '', ''] — messy!

# Rule: Use split() (no args) for whitespace. Use split(delimiter) for specific delimiters.
````

````python
# Limiting splits
print("a,b,c,d".split(",", 2))      # ['a', 'b', 'c,d'] — split at most 2 times

# rsplit() — split from the right
print("a,b,c,d".rsplit(",", 2))     # ['a,b', 'c', 'd'] — useful for paths/URLs

# splitlines() — split by line breaks
text = "line1\nline2\nline3"
print(text.splitlines())              # ['line1', 'line2', 'line3']
````

````python
# join() — called on the SEPARATOR, not the list
print(", ".join(["a", "b", "c"]))          # "a, b, c"
print("-".join(["2026", "04", "12"]))      # "2026-04-12"
print("".join(["h", "e", "l", "l", "o"]))  # "hello"
print("\n".join(["line1", "line2"]))        # Two lines

# join() only works with strings — must convert non-strings
try:
    ", ".join([1, 2, 3])
except TypeError as e:
    print(f"\nTypeError: {e}")

# Fix: convert to strings
print(", ".join(str(x) for x in [1, 2, 3]))  # "1, 2, 3"
````

````python
# split() + join() pattern — extremely common

# Normalize whitespace
messy = "  hello   world  "
clean = " ".join(messy.split())
print(f"Normalized: '{clean}'")

# Convert delimiter
csv_data = "a,b,c"
pipe_data = " | ".join(csv_data.split(","))
print(f"Converted:  '{pipe_data}'")

# Reverse words in a sentence (common interview question!)
sentence = "hello world foo"
reversed_words = " ".join(sentence.split()[::-1])
print(f"Reversed words: '{reversed_words}'")
````

## 11. String Methods — Replacing & Translating

````python
s = "hello world"

# replace(old, new) — replaces ALL occurrences by default
print(s.replace("world", "Python"))   # "hello Python"
print(s.replace("l", "L"))            # "heLLo worLd" — ALL occurrences
print(s.replace("l", "L", 1))         # "heLlo world" — only first occurrence

# Chain replacements
dirty = "a-b_c"
clean = dirty.replace("-", " ").replace("_", " ")
print(f"Chained replace: '{clean}'")
````

````python
# translate() + maketrans() — bulk character replacement (faster than chained replace)

# Map vowels to numbers
table = str.maketrans("aeiou", "12345")
print("hello world".translate(table))    # "h2ll4 w4rld"

# Delete characters
table = str.maketrans("", "", "aeiou")
print("hello world".translate(table))    # "hll wrld"

# Both replace and delete
table = str.maketrans({"a": "1", "e": "2", "l": None})
print("hello".translate(table))          # "h2o"
````

## 12. String Methods — Content Checks (`is*` methods)

All return `True` or `False`. **All return `False` for empty strings.** This is a common gotcha.

````python
# Character type checks
print("=== isalpha() — all alphabetic? ===")
print(f"'hello':     {'hello'.isalpha()}")        # True
print(f"'hello123':  {'hello123'.isalpha()}")     # False
print(f"'hello world': {'hello world'.isalpha()}") # False (space is not alpha)

print("\n=== isdigit() — all digits? ===")
print(f"'123':   {'123'.isdigit()}")     # True
print(f"'12.3':  {'12.3'.isdigit()}")    # False (dot is not a digit)
print(f"'-5':    {'-5'.isdigit()}")       # False (minus is not a digit)

print("\n=== isalnum() — all alphanumeric? ===")
print(f"'hello123': {'hello123'.isalnum()}")   # True
print(f"'hello 123': {'hello 123'.isalnum()}") # False (space)
````

````python
# Whitespace and case checks
print("=== isspace() ===")
print(f"'   ':    {'   '.isspace()}")      # True
print(f"' \\t\\n': {' \t\n'.isspace()}")   # True
print(f"'':       {''.isspace()}")           # False! Empty string is NOT whitespace

print("\n=== Case checks ===")
print(f"'HELLO'.isupper(): {'HELLO'.isupper()}")         # True
print(f"'hello'.islower(): {'hello'.islower()}")         # True
print(f"'Hello World'.istitle(): {'Hello World'.istitle()}") # True
print(f"'Hello world'.istitle(): {'Hello world'.istitle()}") # False (lowercase w)
````

````python
# The empty string gotcha — ALL is* methods return False on ""
empty = ""
print(f"''.isalpha(): {empty.isalpha()}")
print(f"''.isdigit(): {empty.isdigit()}")
print(f"''.isalnum(): {empty.isalnum()}")
print(f"''.isspace(): {empty.isspace()}")
print(f"''.isupper(): {empty.isupper()}")
print(f"''.islower(): {empty.islower()}")
print("ALL False! Remember this for edge cases.")
````

````python
# Numeric vs digit vs decimal — subtle but important differences
print("=== Strictness levels ===")
print(f"'123'.isdecimal(): {'123'.isdecimal()}")   # True — strictest
print(f"'123'.isdigit():   {'123'.isdigit()}")     # True
print(f"'123'.isnumeric(): {'123'.isnumeric()}")   # True — broadest

print(f"\n'²'.isdecimal():  {'²'.isdecimal()}")    # False — superscript not decimal
print(f"'²'.isdigit():    {'²'.isdigit()}")        # True — superscript IS a digit
print(f"'²'.isnumeric():  {'²'.isnumeric()}")      # True

print(f"\n'½'.isdecimal():  {'½'.isdecimal()}")    # False
print(f"'½'.isdigit():    {'½'.isdigit()}")        # False
print(f"'½'.isnumeric():  {'½'.isnumeric()}")      # True — fraction IS numeric

print("\nRule: isdecimal ⊂ isdigit ⊂ isnumeric (from strictest to broadest)")
````

## 14. String Methods — Partition

````python
# partition(sep) — split into 3 parts: (before, sep, after)
print("hello-world-foo".partition("-"))    # ('hello', '-', 'world-foo')
print("hello world".partition("-"))        # ('hello world', '', '') — not found

# rpartition(sep) — from the right
print("hello-world-foo".rpartition("-"))   # ('hello-world', '-', 'foo')

# Practical use: parsing key=value pairs
line = "name=Sevan"
key, _, value = line.partition("=")
print(f"\nKey: '{key}', Value: '{value}'")

# Practical use: splitting email
email = "user@example.com"
username, _, domain = email.partition("@")
print(f"Username: '{username}', Domain: '{domain}'")
````

## 15. String Formatting

Three ways to format strings in Python. **F-strings are preferred** for modern code.

````python
# F-strings (recommended — Python 3.6+)
name = "Sevan"
age = 25

# Basic interpolation
print(f"Hello, {name}!")
print(f"Next year: {age + 1}")      # Expressions work
print(f"Name length: {len(name)}")  # Function calls work
print(f"Uppercase: {name.upper()}") # Methods work
````

````python
# F-string number formatting
pi = 3.14159

print(f"2 decimals:    {pi:.2f}")
print(f"Comma sep:     {1000000:,}")
print(f"Percentage:    {0.75:.0%}")
print(f"Zero-padded:   {42:05d}")

# Alignment in f-strings
print(f"Right align:   {'hello':>20}")
print(f"Left align:    {'hello':<20}")
print(f"Center:        {'hello':^20}")
print(f"Center fill:   {'hello':*^20}")
````

````python
# Debug format (Python 3.8+) — extremely useful for debugging
x = 42
name = "Sevan"
data = [1, 2, 3]

print(f"{x=}")
print(f"{name=}")
print(f"{len(data)=}")
print(f"{data[0] + data[1]=}")
````

````python
# .format() method — know this for reading older code
print("Hello, {}!".format("Sevan"))
print("{0} and {1}, then {0} again".format("a", "b"))
print("{name} is {age}".format(name="Sevan", age=25))
print("{:.2f}".format(3.14159))

# %-formatting (legacy — know it for reading very old code)
print("Hello, %s! You are %d." % ("Sevan", 25))
print("Pi is %.2f" % 3.14159)
````

````python
# Manual character classification using ord()
# This is how many interview solutions work under the hood

def is_digit(c):
    """Check if character is a digit without using .isdigit()"""
    return '0' <= c <= '9'

def is_lower(c):
    """Check if character is lowercase without using .islower()"""
    return 'a' <= c <= 'z'

def is_upper(c):
    """Check if character is uppercase without using .isupper()"""
    return 'A' <= c <= 'Z'

def to_upper(c):
    """Convert lowercase to uppercase manually"""
    if 'a' <= c <= 'z':
        return chr(ord(c) - 32)  # Distance between 'a' and 'A' is 32
    return c

def to_lower(c):
    """Convert uppercase to lowercase manually"""
    if 'A' <= c <= 'Z':
        return chr(ord(c) + 32)
    return c

# Test them
print(f"is_digit('5'): {is_digit('5')}")
print(f"is_digit('a'): {is_digit('a')}")
print(f"to_upper('a'): '{to_upper('a')}'")
print(f"to_lower('Z'): '{to_lower('Z')}'")
````

## 18. String Traversal — The Foundation of String Problems

Nearly every LeetCode string problem requires some form of traversal. Mastering different traversal techniques is **essential**. This section goes deep because traversal is the single most important skill for string interview problems.

### Traversal Types You Must Know:
1. Simple forward traversal (`for char in s`)
2. Index-based traversal (`for i in range(len(s))`)
3. Enumerated traversal (`for i, char in enumerate(s)`)
4. Reverse traversal
5. Two-pointer traversal
6. Sliding window traversal
7. Character-by-character with state
8. Traversal with frequency counting
9. Traversal with building a new string
10. Multi-string parallel traversal

### 18.1 Simple Forward Traversal

The most basic pattern. Iterate over each character one at a time.

````python
# Basic character-by-character traversal
s = "hello"
for char in s:
    print(char, end=" ")
print()

# Use case: Count specific characters
text = "Hello World"
vowel_count = 0
for char in text:
    if char.lower() in "aeiou":
        vowel_count += 1
print(f"Vowels in '{text}': {vowel_count}")

# Use case: Build a new string by filtering
text = "h3ll0 w0rld"
letters_only = ""
for char in text:
    if char.isalpha():
        letters_only += char
print(f"Letters only: '{letters_only}'")

# Better way (using join + generator):
letters_only = "".join(char for char in text if char.isalpha())
print(f"Letters only (join): '{letters_only}'")
````

### 18.2 Index-Based Traversal

When you need to know the **position** of each character, or need to look ahead/behind.

````python
# Index-based traversal with range(len(s))
s = "Python"
for i in range(len(s)):
    print(f"Index {i}: '{s[i]}'")

# When you need to look ahead (compare current with next)
s = "aabbcc"
print(f"\nAdjacent pairs in '{s}':")
for i in range(len(s) - 1):  # Stop one before the end!
    print(f"  s[{i}]='{s[i]}' and s[{i+1}]='{s[i+1]}' — {'same' if s[i] == s[i+1] else 'different'}")
````

````python
# Index-based traversal: Run-Length Encoding
# This is a CLASSIC interview problem that requires looking ahead
s = "aaabbcccc"
result = []
i = 0
while i < len(s):
    count = 1
    # Keep moving forward while next char is the same
    while i + count < len(s) and s[i + count] == s[i]:
        count += 1
    result.append(f"{s[i]}{count}")
    i += count  # Jump past the group

compressed = "".join(result)
print(f"'{s}' compressed to '{compressed}'")
````

### 18.3 Enumerated Traversal

`enumerate()` gives you both the index AND the character. This is the **preferred** way when you need both.

````python
s = "hello"
for i, char in enumerate(s):
    print(f"Index {i}: '{char}'")

# Use case: Find all positions of a character
text = "banana"
target = "a"
positions = []
for i, char in enumerate(text):
    if char == target:
        positions.append(i)
print(f"\n'{target}' found at positions {positions} in '{text}'")

# Same thing with list comprehension:
positions = [i for i, c in enumerate(text) if c == target]
print(f"Same result: {positions}")
````

````python
# Use case: First non-repeating character (classic interview question)
from collections import Counter

def first_unique_char(s):
    """Find the index of the first non-repeating character. Return -1 if none."""
    freq = Counter(s)
    for i, char in enumerate(s):
        if freq[char] == 1:
            return i
    return -1

# This requires TWO passes: one to count, one to find
# Pass 1: Counter(s) traverses the whole string → O(n)
# Pass 2: enumerate loop → O(n)
# Total: O(n) time, O(1) space (at most 26 letters)

test_cases = ["leetcode", "loveleetcode", "aabb"]
for s in test_cases:
    idx = first_unique_char(s)
    char = s[idx] if idx != -1 else "none"
    print(f"first_unique_char('{s}') = {idx} (char='{char}')")
````

### 18.4 Reverse Traversal

Going through a string backwards. There are several ways to do this.

````python
s = "hello"

# Method 1: Slicing with [::-1]
print(f"Reversed slice: '{s[::-1]}'")

# Method 2: reversed() built-in
print(f"reversed():     '{''.join(reversed(s))}'")

# Method 3: Index-based reverse traversal
print("Index-based: ", end="")
for i in range(len(s) - 1, -1, -1):  # From last index to 0
    print(s[i], end="")
print()

# Method 4: Negative indexing
print("Negative idx: ", end="")
for i in range(-1, -len(s) - 1, -1):
    print(s[i], end="")
print()
````

````python
# Reverse traversal use case: Check if a string ends with a pattern
# (without using endswith)

def ends_with(s, suffix):
    """Check if s ends with suffix by traversing backwards."""
    if len(suffix) > len(s):
        return False
    for i in range(1, len(suffix) + 1):
        if s[-i] != suffix[-i]:
            return False
    return True

print(f"ends_with('hello', 'llo'): {ends_with('hello', 'llo')}")
print(f"ends_with('hello', 'xyz'): {ends_with('hello', 'xyz')}")

# Reverse traversal use case: Reverse words in a string
sentence = "the sky is blue"
words = sentence.split()
reversed_sentence = " ".join(words[::-1])
print(f"\nReversed words: '{reversed_sentence}'")
````

### 18.5 Two-Pointer Traversal

One of the most powerful and frequently tested patterns. Two pointers start at different positions and move toward each other or in the same direction.

**Common interview problems using two pointers on strings:**
- Palindrome check
- Reverse a string in place (on a list of chars)
- Remove duplicates
- Valid palindrome with non-alphanumeric chars

````python
# Two-pointer palindrome check
# Left pointer starts at beginning, right pointer starts at end
# They move toward each other

def is_palindrome(s):
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

test_words = ["racecar", "hello", "madam", "a", ""]
for word in test_words:
    print(f"is_palindrome('{word}'): {is_palindrome(word)}")
````

````python
# Two-pointer: Valid Palindrome (LeetCode #125)
# Ignore non-alphanumeric characters and case

def is_valid_palindrome(s):
    left, right = 0, len(s) - 1
    while left < right:
        # Skip non-alphanumeric from left
        while left < right and not s[left].isalnum():
            left += 1
        # Skip non-alphanumeric from right
        while left < right and not s[right].isalnum():
            right -= 1
        # Compare (case-insensitive)
        if s[left].lower() != s[right].lower():
            return False
        left += 1
        right -= 1
    return True

test_cases = [
    "A man, a plan, a canal: Panama",
    "race a car",
    " ",
    "Was it a car or a cat I saw?"
]
for s in test_cases:
    print(f"is_valid_palindrome('{s}'): {is_valid_palindrome(s)}")
````

````python
# Two-pointer: Reverse a list of characters in-place
# (Strings are immutable, so we work with a list of chars)

def reverse_string(chars):
    """Reverse a list of characters in-place using two pointers."""
    left, right = 0, len(chars) - 1
    while left < right:
        chars[left], chars[right] = chars[right], chars[left]
        left += 1
        right -= 1

chars = list("hello")
print(f"Before: {chars}")
reverse_string(chars)
print(f"After:  {chars}")
print(f"As string: '{''.join(chars)}'")
````

### 18.6 Sliding Window Traversal

A window of fixed or variable size slides across the string. This is one of the most important patterns for medium/hard interview problems.

**Common problems:**
- Longest substring without repeating characters
- Minimum window substring
- Find all anagrams in a string
- Longest repeating character replacement

````python
# Fixed-size sliding window: Check all substrings of length k
s = "abcdefgh"
k = 3

print(f"All substrings of length {k} in '{s}':")
for i in range(len(s) - k + 1):
    window = s[i:i+k]
    print(f"  [{i}:{i+k}] = '{window}'")
````

````python
# Variable-size sliding window: Longest substring without repeating characters
# (LeetCode #3 — one of the most popular interview questions)

def length_of_longest_substring(s):
    """Find the length of the longest substring without repeating characters."""
    char_set = set()
    left = 0
    max_len = 0
    
    for right in range(len(s)):
        # If we find a duplicate, shrink window from left
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        # Add current character
        char_set.add(s[right])
        # Update max length
        max_len = max(max_len, right - left + 1)
    
    return max_len

test_cases = ["abcabcbb", "bbbbb", "pwwkew", "", "abcdef"]
for s in test_cases:
    print(f"longest_substring('{s}'): {length_of_longest_substring(s)}")
````

````python
# Sliding window: Find all anagrams of pattern in text
# Return starting indices of all anagram windows

from collections import Counter

def find_anagrams(s, p):
    """Find all start indices where an anagram of p begins in s."""
    if len(p) > len(s):
        return []
    
    p_count = Counter(p)
    window_count = Counter(s[:len(p)])
    result = []
    
    if window_count == p_count:
        result.append(0)
    
    # Slide the window
    for i in range(len(p), len(s)):
        # Add new character to window
        window_count[s[i]] += 1
        # Remove old character from window
        old_char = s[i - len(p)]
        window_count[old_char] -= 1
        if window_count[old_char] == 0:
            del window_count[old_char]
        # Check if current window is an anagram
        if window_count == p_count:
            result.append(i - len(p) + 1)
    
    return result

print(f"find_anagrams('cbaebabacd', 'abc'): {find_anagrams('cbaebabacd', 'abc')}")
print(f"find_anagrams('abab', 'ab'): {find_anagrams('abab', 'ab')}")
````

### 18.7 Character-by-Character with State

Traversing while keeping track of some running state. This pattern appears in many problems.

````python
# Building a frequency map (most fundamental pattern)
s = "hello world"

# Method 1: Manual dictionary
freq = {}
for char in s:
    freq[char] = freq.get(char, 0) + 1
print(f"Manual freq: {freq}")

# Method 2: defaultdict
from collections import defaultdict
freq2 = defaultdict(int)
for char in s:
    freq2[char] += 1
print(f"defaultdict: {dict(freq2)}")

# Method 3: Counter (most Pythonic)
from collections import Counter
freq3 = Counter(s)
print(f"Counter:     {dict(freq3)}")
print(f"Most common: {freq3.most_common(3)}")
````

````python
# Traversal with state: String compression (run-length encoding)
# "aaabbbccdd" → "a3b3c2d2"

def compress(s):
    if not s:
        return ""
    
    result = []
    current_char = s[0]
    count = 1
    
    for i in range(1, len(s)):  # Start from index 1
        if s[i] == current_char:
            count += 1
        else:
            result.append(current_char + str(count))
            current_char = s[i]
            count = 1
    
    # Don't forget the last group!
    result.append(current_char + str(count))
    return "".join(result)

test_cases = ["aaabbbccdd", "abcd", "aaa", "", "aabbcc"]
for s in test_cases:
    print(f"compress('{s}'): '{compress(s)}'")
````

````python
# Traversal with state: Valid parentheses using a stack
# This combines traversal with a stack data structure

def is_valid_parens(s):
    stack = []
    matching = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in matching:  # It's a closing bracket
            if not stack or stack[-1] != matching[char]:
                return False
            stack.pop()
        else:  # It's an opening bracket
            stack.append(char)
    
    return len(stack) == 0

tests = ["()", "()[]{}", "(]", "([)]", "{[]}", ""]
for s in tests:
    print(f"is_valid_parens('{s}'): {is_valid_parens(s)}")
````

### 18.8 Traversal with Building a New String

Many problems require building a new string character by character as you traverse. Always collect into a **list** and `.join()` at the end for O(n) performance.

````python
# Pattern: Filter and transform characters
s = "Hello, World! 123"

# Keep only letters, convert to lowercase
cleaned = []
for char in s:
    if char.isalpha():
        cleaned.append(char.lower())
result = "".join(cleaned)
print(f"Filtered: '{result}'")

# Same thing with list comprehension (more Pythonic):
result = "".join(c.lower() for c in s if c.isalpha())
print(f"Comprehension: '{result}'")
````

````python
# Caesar cipher — shift each letter by n positions
# This is a traversal + transformation problem

def caesar_cipher(text, shift):
    result = []
    for char in text:
        if char.isalpha():
            # Determine base ('A' for uppercase, 'a' for lowercase)
            base = ord('A') if char.isupper() else ord('a')
            # Shift and wrap around using modulo
            shifted = chr((ord(char) - base + shift) % 26 + base)
            result.append(shifted)
        else:
            result.append(char)  # Non-alpha characters stay
    return "".join(result)

print(f"caesar('abc', 1):          '{caesar_cipher('abc', 1)}'")
print(f"caesar('xyz', 3):          '{caesar_cipher('xyz', 3)}'")
print(f"caesar('Hello, World!', 13): '{caesar_cipher('Hello, World!', 13)}'")
print(f"ROT13 back:                '{caesar_cipher('Uryyb, Jbeyq!', 13)}'")
````

### 18.9 Multi-String Parallel Traversal

Traversing two or more strings simultaneously. Used in problems like:
- Comparing strings
- Merging strings
- Longest common prefix
- Add binary strings

````python
# zip() for parallel traversal of same-length strings
s1 = "hello"
s2 = "world"
for c1, c2 in zip(s1, s2):
    print(f"({c1}, {c2})", end=" ")
print()

# Count matching positions
matches = sum(1 for c1, c2 in zip(s1, s2) if c1 == c2)
print(f"Characters matching at same position: {matches}")
````

````python
# Longest common prefix (LeetCode #14)
def longest_common_prefix(strs):
    if not strs:
        return ""
    
    # Use the shortest string as the limit
    prefix = []
    for i in range(len(min(strs, key=len))):
        # Check if all strings have the same character at position i
        char = strs[0][i]
        if all(s[i] == char for s in strs):
            prefix.append(char)
        else:
            break
    return "".join(prefix)

print(f"LCP of ['flower','flow','flight']: '{longest_common_prefix(['flower','flow','flight'])}'")
print(f"LCP of ['dog','racecar','car']:    '{longest_common_prefix(['dog','racecar','car'])}'")
print(f"LCP of ['abc','abc','abc']:        '{longest_common_prefix(['abc','abc','abc'])}'")
````

````python
# Add binary strings (LeetCode #67)
# Traverse both strings from right to left, like manual addition

def add_binary(a, b):
    result = []
    carry = 0
    i, j = len(a) - 1, len(b) - 1
    
    while i >= 0 or j >= 0 or carry:
        total = carry
        if i >= 0:
            total += int(a[i])
            i -= 1
        if j >= 0:
            total += int(b[j])
            j -= 1
        result.append(str(total % 2))
        carry = total // 2
    
    return "".join(reversed(result))

print(f"add_binary('11', '1'):    '{add_binary('11', '1')}'")
print(f"add_binary('1010', '1011'): '{add_binary('1010', '1011')}'")
````

### 18.10 Traversal with Frequency Counting

Combining traversal with a hash map (dictionary) to count character frequencies. This pattern solves a huge number of string problems.

````python
# Anagram check — compare frequency maps
from collections import Counter

def is_anagram(s1, s2):
    return Counter(s1.lower()) == Counter(s2.lower())

# Alternatively without Counter (interview might ask this):
def is_anagram_manual(s1, s2):
    if len(s1) != len(s2):
        return False
    freq = {}
    for c in s1.lower():
        freq[c] = freq.get(c, 0) + 1
    for c in s2.lower():
        freq[c] = freq.get(c, 0) - 1
        if freq[c] < 0:
            return False
    return True

pairs = [("listen", "silent"), ("hello", "world"), ("Astronomer", "Moon starer")]
for s1, s2 in pairs:
    print(f"is_anagram('{s1}', '{s2}'): {is_anagram(s1.replace(' ', ''), s2.replace(' ', ''))}")
````

````python
# Group anagrams (LeetCode #49)
# Use sorted string as a hash key

from collections import defaultdict

def group_anagrams(strs):
    groups = defaultdict(list)
    for s in strs:
        # Use sorted characters as the key — all anagrams sort to the same string
        key = "".join(sorted(s))
        groups[key].append(s)
    return list(groups.values())

words = ["eat", "tea", "tan", "ate", "nat", "bat"]
result = group_anagrams(words)
print(f"Input:  {words}")
print(f"Groups: {result}")

# Time: O(n * k log k) where n = number of words, k = max word length
# Space: O(n * k)
````

## 19. Common Interview Patterns Summary

These are the string patterns that appear most frequently in coding interviews. Each one builds on the traversal techniques above.

````python
# Pattern 1: Reverse a string
s = "hello"
print(f"Slice:    '{s[::-1]}'")
print(f"Join:     '{''.join(reversed(s))}'")

# Pattern 2: Palindrome check
word = "racecar"
print(f"\n'{word}' palindrome: {word == word[::-1]}")

# Pattern 3: Clean palindrome (LeetCode style)
def clean_palindrome(s):
    cleaned = "".join(c.lower() for c in s if c.isalnum())
    return cleaned == cleaned[::-1]

print(f"clean palindrome 'A man, a plan, a canal: Panama': {clean_palindrome('A man, a plan, a canal: Panama')}")
````

````python
# Pattern 4: Character frequency → find most common / least common / unique
from collections import Counter

s = "abracadabra"
freq = Counter(s)
print(f"Frequency of '{s}': {dict(freq)}")
print(f"Most common: {freq.most_common(1)}")
print(f"Least common: {freq.most_common()[-1]}")
print(f"Unique chars: {[c for c, count in freq.items() if count == 1]}")
````

````python
# Pattern 5: Reverse words in a string
def reverse_words(s):
    return " ".join(s.split()[::-1])

print(f"reverse_words('hello world foo'): '{reverse_words('hello world foo')}'")
print(f"reverse_words('  hello   world  '): '{reverse_words('  hello   world  ')}'")

# Pattern 6: String compression
def compress_smart(s):
    """Only compress if result is shorter."""
    result = []
    i = 0
    while i < len(s):
        count = 1
        while i + count < len(s) and s[i + count] == s[i]:
            count += 1
        result.append(s[i] + str(count))
        i += count
    compressed = "".join(result)
    return compressed if len(compressed) < len(s) else s

print(f"\ncompress('aaabbbccc'): '{compress_smart('aaabbbccc')}'")
print(f"compress('abc'): '{compress_smart('abc')}'")
````

## 20. Edge Cases to Watch For

These are the edge cases that appear in interviews and that cause bugs in production. Run each cell to internalize the behavior.

````python
# Empty string edge cases
print("=== Empty String ===")
print(f"len(''):       {len('')}")
print(f"'' == '':      {'' == ''}")
print(f"bool(''):      {bool('')}")         # False — falsy
print(f"''.split():    {''.split()}")
print(f"''.split(','):  {''.split(',')}")    # [''] NOT [] !!!
print(f"','.join([]):  {','.join([])}")
print(f"'' in 'hello': {'' in 'hello'}")
print(f"''.isalpha():  {''.isalpha()}")
````

````python
# Single character edge cases
print("=== Single Character ===")
print(f"'a'[0]:     '{('a')[0]}'")
print(f"'a'[::-1]:  '{'a'[::-1]}'")
print(f"len('a'):   {len('a')}")
print(f"'a' == 'a': {'a' == 'a'}")

# Whitespace-only strings
print("\n=== Whitespace Only ===")
print(f"'   '.strip():   '{'   '.strip()}'")
print(f"'   '.split():   {'   '.split()}")
print(f"'   '.isspace(): {'   '.isspace()}")
print(f"bool('   '):     {bool('   ')}")   # True! Space is not empty
````

````python
# Numeric string gotchas
print("=== Numeric Strings ===")
print(f"'10' < '9':   {'10' < '9'}")      # True! Lexicographic, not numeric
print(f"'10' < '2':   {'10' < '2'}")      # True! '1' < '2'
print(f"int('10') < int('9'): {int('10') < int('9')}")  # False (correct numeric)

# String multiplication edge cases
print(f"\n'abc' * 0:  '{'abc' * 0}'")
print(f"'abc' * -1: '{'abc' * -1}'")

# None gotcha
print(f"\nstr(None):    '{str(None)}'")
print(f"'' is not None: {'' is not None}")
````

````python
# Unicode edge cases
print("=== Unicode ===")
print(f"len('café'):  {len('café')}")
print(f"'café'[3]:    '{'café'[3]}'")
emoji = "🐍"
print(f"len('🐍'):    {len(emoji)}")
print(f"'🐍' is valid: {True}")
````

````python
# SQL SAFETY — never use f-strings for SQL queries!

name = "Alice"

# DANGEROUS — SQL injection vulnerability:
bad_query = f"SELECT * FROM users WHERE name = '{name}'"
print(f"BAD (vulnerable):  {bad_query}")

# What if name = "'; DROP TABLE users; --" ?
evil_name = "'; DROP TABLE users; --"
bad_query = f"SELECT * FROM users WHERE name = '{evil_name}'"
print(f"SQL INJECTION:     {bad_query}")

# SAFE — use parameterized queries (shown as pseudocode here):
print(f"\nSAFE: cursor.execute('SELECT * FROM users WHERE name = %s', (name,))")
print("Always use parameterized queries in production!")
````

## 22. Performance Reference

| Operation | Time Complexity | Notes |
|-----------|----------------|-------|
| `s[i]` indexing | O(1) | Direct access |
| `s[a:b]` slicing | O(b-a) | Creates new string |
| `len(s)` | O(1) | Stored internally |
| `s + t` concat | O(len(s) + len(t)) | Creates new string |
| `s += t` in loop (n times) | O(n²) total | **Avoid!** |
| `"".join(list)` | O(total length) | **Use this instead** |
| `s.find(sub)` / `s.index(sub)` | O(n * m) worst case | n=len(s), m=len(sub) |
| `s.replace(old, new)` | O(n) | Scans entire string |
| `s.split()` | O(n) | Linear scan |
| `sorted(s)` | O(n log n) | Returns list |
| `s in other` | O(n * m) worst case | Substring search |
| `Counter(s)` | O(n) | Single pass |

### Key Takeaway
The biggest performance trap is **string concatenation in a loop**. Always use a list + `.join()`.
