## Level 1 ⭐ — Basics

```python
# Drill 1: Return the first and last character of a string concatenated together.
# "hello" → "ho"   |   "a" → "aa"   |   "" → ""

def first_and_last(s: str) -> str:
    # TODO
    if s == '':
        return ''
    return s[0] + s[-1]

assert first_and_last("hello") == "ho"
assert first_and_last("a") == "aa"
assert first_and_last("") == ""
assert first_and_last("ab") == "ab"
assert first_and_last("Python") == "Pn"
print("Drill 1 passed ✓")
```

```python
# Drill 2: Return a string with the first character uppercased and the rest lowercased.
# "hELLO" → "Hello"   |   "" → ""   |   "a" → "A"

def capitalize_first(s: str) -> str:
    # TODO
    if s == '':
        return ''
    first = s[0].upper()
    rest = s[1:].lower()
    return first + rest
    

assert capitalize_first("hELLO") == "Hello"
assert capitalize_first("") == ""
assert capitalize_first("a") == "A"
assert capitalize_first("PYTHON") == "Python"
assert capitalize_first("already") == "Already"
print("Drill 2 passed ✓")
```

```python
# Drill 3: Return every other character starting from index 0.
# "abcdef" → "ace"   |   "hello" → "hlo"   |   "" → ""

def every_other(s: str) -> str:
    # TODO
    return s[::2]

assert every_other("abcdef") == "ace"
assert every_other("hello") == "hlo"
assert every_other("") == ""
assert every_other("a") == "a"
assert every_other("ab") == "a"
print("Drill 3 passed ✓")
```

## Level 2 ⭐⭐ — Methods & Formatting

```python
# Drill 4: Given a sentence, return the number of words.
# Words are separated by whitespace. Handle multiple spaces and leading/trailing spaces.
# "  hello   world  " → 2   |   "" → 0   |   "one" → 1

def word_count(s: str) -> int:
    # TODO
    return len(s.split())

assert word_count("hello world") == 2
assert word_count("  hello   world  ") == 2
assert word_count("") == 0
assert word_count("   ") == 0
assert word_count("one") == 1
assert word_count("a b c d e") == 5
print("Drill 4 passed ✓")
```

```python
# Drill 5: Convert a string to "snake_case".
# Replace all spaces with underscores and lowercase everything.
# Strip leading/trailing spaces first.
# "Hello World" → "hello_world"   |   "  Already snake  " → "already_snake"

def to_snake_case(s: str) -> str:
    # TODO
    return s.lstrip().rstrip().replace(" ", "_").lower()

assert to_snake_case("Hello World") == "hello_world"
assert to_snake_case("  Already snake  ") == "already_snake"
assert to_snake_case("ALL CAPS HERE") == "all_caps_here"
assert to_snake_case("single") == "single"
assert to_snake_case("") == ""
print("Drill 5 passed ✓")
```

```python
# Drill 6: Build a greeting string using f-string formatting.
# Given name and age, return: "Hello, my name is {name} and I am {age} years old."

def greeting(name: str, age: int) -> str:
    # TODO
    return f'Hello, my name is {name} and I am {age} years old.'

assert greeting("Alice", 30) == "Hello, my name is Alice and I am 30 years old."
assert greeting("Bob", 0) == "Hello, my name is Bob and I am 0 years old."
assert greeting("", 25) == "Hello, my name is  and I am 25 years old."
print("Drill 6 passed ✓")
```

```python
# Drill 7: Join a list of words with a separator, but skip empty strings.
# join_non_empty(["a", "", "b", "", "c"], "-") → "a-b-c"

def join_non_empty(words: list, sep: str) -> str:
    # TODO
    non_empty = [word for word in words if word != '']
    return sep.join(non_empty)

assert join_non_empty(["a", "", "b", "", "c"], "-") == "a-b-c"
assert join_non_empty(["", "", ""], ",") == ""
assert join_non_empty(["hello", "world"], " ") == "hello world"
assert join_non_empty([], "-") == ""
assert join_non_empty(["only"], "|") == "only"
print("Drill 7 passed ✓")
```

## Level 3 ⭐⭐⭐ — Multi-Step Transformations

```python
# Drill 8: Mask all but the last 4 characters of a string with '*'.
# "1234567890" → "******7890"   |   "ab" → "ab"   |   "" → ""
# If the string has 4 or fewer characters, return it unchanged.

def mask_string(s: str) -> str:
    # TODO
    if len(s) <= 4:
        return s
    
    return ('*' * (len(s)-4)) + s[len(s)-4:]

assert mask_string("1234567890") == "******7890"
assert mask_string("abcd") == "abcd"
assert mask_string("ab") == "ab"
assert mask_string("") == ""
assert mask_string("hello world") == "*******orld"
assert mask_string("12345") == "*2345"
print("Drill 8 passed ✓")
```

```python
# Drill 9: Count how many times a substring appears (case-insensitive).
# count_substring("Hello hello HELLO", "hello") → 3

def count_substring(text: str, sub: str) -> int:
    # TODO
    return text.lower().count(sub.lower())


assert count_substring("Hello hello HELLO", "hello") == 3
assert count_substring("aaa", "a") == 3
assert count_substring("abc", "z") == 0
assert count_substring("", "a") == 0
assert count_substring("Mississippi", "ss") == 2
assert count_substring("Mississippi", "SS") == 2
print("Drill 9 passed ✓")
```

```python
# Drill 10: Extract the domain from an email address.
# "user@example.com" → "example.com"
# If there's no '@', return "".
# If there are multiple '@', use the LAST one (technically valid edge case).

def extract_domain(email: str) -> str:
    # TODO
    if '@' not in email:
        return ''
    
    return email[email.rfind('@')+1:]


assert extract_domain("user@example.com") == "example.com"
assert extract_domain("no-at-sign") == ""
assert extract_domain("weird@@double.com") == "double.com"
assert extract_domain("a@b") == "b"
assert extract_domain("") == ""
print("Drill 10 passed ✓")
```

```python
# Drill 11: Remove all vowels from a string (both cases).
# "Hello World" → "Hll Wrld"

def remove_vowels(s: str) -> str:
    # TODO
    vowels = 'aeiouAEIOU'
    return ''.join(char for char in s if char not in vowels)

assert remove_vowels("Hello World") == "Hll Wrld"
assert remove_vowels("aeiouAEIOU") == ""
assert remove_vowels("bcdfg") == "bcdfg"
assert remove_vowels("") == ""
assert remove_vowels("Python") == "Pythn"
print("Drill 11 passed ✓")
```

## Level 4 ⭐⭐⭐⭐ — Interview-Style Problems

```python
# Drill 12: Check if two strings are anagrams of each other.
# Ignore case and spaces.
# "listen" and "silent" → True
# "Astronomer" and "Moon starer" → True

def are_anagrams(s1: str, s2: str) -> bool:
    # TODO
    
    if len(s1.lower().replace(" ", "")) != len(s2.lower().replace(" ", '')):
        return False
    
    sorted_s1 = ''.join(sorted(s1.replace(" ", '').lower()))
    sorted_s2 = ''.join(sorted(s2.replace(" ", '').lower()))

    return sorted_s1 == sorted_s2

assert are_anagrams("listen", "silent") == True
assert are_anagrams("Astronomer", "Moon starer") == True
assert are_anagrams("hello", "world") == False
assert are_anagrams("", "") == True
assert are_anagrams("a", "a") == True
assert are_anagrams("ab", "a") == False
assert are_anagrams("Dormitory", "dirty room") == True
print("Drill 12 passed ✓")
```

```python
# Drill 13: Check if a string has all unique characters (no repeats).
# Case-sensitive: 'a' and 'A' are different.

def all_unique(s: str) -> bool:
    # TODO
    if len(s) == 0 or len(s) == 1:
        return True
    
    seen = set()
    for val in s:
        if val in seen:
            return False
        seen.add(val)

    return True


assert all_unique("abcdef") == True
assert all_unique("aAbBcC") == True  # case-sensitive
assert all_unique("hello") == False
assert all_unique("") == True
assert all_unique("a") == True
assert all_unique("abcda") == False
print("Drill 13 passed ✓")
```

```python
# Drill 14: Given a string of words, return the longest word.
# If there's a tie, return the FIRST one.
# "The quick brown fox" → "quick"

def longest_word(s: str) -> str:
    # TODO
    vals = s.split()
    longest = str()   
    
    for i in vals:
        if len(i) > len(longest):
            longest = i
    
    return longest

assert longest_word("The quick brown fox") == "quick"
assert longest_word("a bb ccc dd") == "ccc"
assert longest_word("single") == "single"
assert longest_word("ab cd") == "ab"  # tie → first
assert longest_word("") == ""
assert longest_word("I am a student") == "student"
print("Drill 14 passed ✓")
```

```python
# Drill 15: Caesar cipher — shift each letter by n positions.
# Wrap around: 'z' shifted by 1 → 'a'. Preserve case. Keep non-alpha chars unchanged.
# caesar("abc", 1) → "bcd"   |   caesar("xyz", 3) → "abc"

def caesar(text: str, shift: int) -> str:
    result = []

    for ch in text:
        if ch.isalpha():
            base = ord('a') if ch.islower() else ord('A')
            new_pos = (ord(ch) - base + shift) % 26
            result.append(chr(new_pos + base))
        else:
            result.append(ch)

    return "".join(result)


assert caesar("abc", 1) == "bcd"
assert caesar("xyz", 3) == "abc"
assert caesar("Hello, World!", 13) == "Uryyb, Jbeyq!"
assert caesar("abc", 0) == "abc"
assert caesar("ABC", 26) == "ABC"  # full rotation
assert caesar("", 5) == ""
assert caesar("abc", -1) == "zab"  # negative shift
print("Drill 15 passed ✓")
```
