## Level 1 ⭐ — Core Fluency

```python
# Drill 1: Reverse a String
# Return the reverse of the input string. Do NOT use reversed() or [::-1].
# Use a loop or manual approach.
# "hello" → "olleh"   |   "" → ""

def reverse_string(s: str) -> str:
    # TODO
    result = ""

    for ch in s:
        result = ch + result
    return result

assert reverse_string("hello") == "olleh"
assert reverse_string("") == ""
assert reverse_string("a") == "a"
assert reverse_string("ab") == "ba"
assert reverse_string("racecar") == "racecar"
print("Drill 1 passed ✓")
```

```python
# Drill 2: Check Palindrome (simple)
# Return True if the string reads the same forwards and backwards.
# Case-sensitive, no stripping. "" is a palindrome.
# "racecar" → True   |   "hello" → False

def is_palindrome_simple(s: str) -> bool:
    # TODO
    return s == s[::-1]
assert is_palindrome_simple("racecar") == True
assert is_palindrome_simple("hello") == False
assert is_palindrome_simple("") == True
assert is_palindrome_simple("a") == True
assert is_palindrome_simple("aba") == True
assert is_palindrome_simple("ab") == False
print("Drill 2 passed ✓")
```

```python
# Drill 3: Count Character Frequency
# Return a dict mapping each character to its count.
# "aabbc" → {"a": 2, "b": 2, "c": 1}   |   "" → {}
# Do NOT use collections.Counter — build the dict yourself.

def char_frequency(s: str) -> dict:
    # TODO
    freq = {}

    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    
    return freq

assert char_frequency("aabbc") == {"a": 2, "b": 2, "c": 1}
assert char_frequency("") == {}
assert char_frequency("x") == {"x": 1}
assert char_frequency("aaa") == {"a": 3}
assert char_frequency("abab") == {"a": 2, "b": 2}
print("Drill 3 passed ✓")
```

```python
# Drill 4: First Non-Repeating Character
# Return the INDEX of the first character that doesn't repeat.
# Return -1 if every character repeats.
# "leetcode" → 0  |  "loveleetcode" → 2  |  "aabb" → -1
# This is LC 387. Build a frequency map, then scan again.

def first_uniq_char(s: str) -> int:
    # TODO
    freq = {}

    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    
    for i,ch in enumerate(s):
        if freq[ch] == 1:
            return i
    return -1


assert first_uniq_char("leetcode") == 0
assert first_uniq_char("loveleetcode") == 2
assert first_uniq_char("aabb") == -1
assert first_uniq_char("") == -1
assert first_uniq_char("a") == 0
assert first_uniq_char("aadadaad") == -1
print("Drill 4 passed ✓")
```

```python
# Drill 5: Reverse Words in a String
# Given a string with words separated by spaces, reverse the order of words.
# Strip leading/trailing spaces. Collapse multiple spaces to one.
# "  the sky  is blue  " → "blue is sky the"
# This is LC 151.

def reverse_words(s: str) -> str:
    # TODO
    clean = s.strip().split()

    return ' '.join(clean[::-1])

assert reverse_words("the sky is blue") == "blue is sky the"
assert reverse_words("  hello world  ") == "world hello"
assert reverse_words("a") == "a"
assert reverse_words("  a  b  c  ") == "c b a"
assert reverse_words("") == ""
print("Drill 5 passed ✓")
```

```python
# Drill 6: String Compression
# Compress consecutive repeated chars: "aaabbccccd" → "a3b2c4d1"
# If the compressed string is NOT shorter, return the original.
# "abc" → "abc" (compressed would be "a1b1c1", which is longer)
# This is a classic CTCI interview problem.

def compress(s: str) -> str:
    # TODO
    result = []
    freq = {}

    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    
    for ch in s:
        result.append(ch)
        result.append(str(freq[ch]))
    
    return ''. join(result)

assert compress("aaabbccccd") == "a3b2c4d1"
assert compress("abc") == "abc"
assert compress("aab") == "aab"  # "a2b1" is same length, return original
assert compress("") == ""
assert compress("aaaa") == "a4"
assert compress("aabbaabb") == "aabbaabb"  # compressed "a2b2a2b2" is same length
print("Drill 6 passed ✓")
```
