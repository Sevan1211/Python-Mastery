## Section 4: Build from Scratch

Write these functions entirely on your own.

```python
# BUILD: Write a function that takes any value and returns its type as a string.
# Example: get_type_name(42) → "int"
# Example: get_type_name("hi") → "str"
# Example: get_type_name(None) → "NoneType"

def get_type_name(value):
    return type(value).__name__

# Tests — all should pass
assert get_type_name(42) == "int"
assert get_type_name(3.14) == "float"
assert get_type_name("hello") == "str"
assert get_type_name(True) == "bool"
assert get_type_name(None) == "NoneType"
print("All tests passed ✓")
```

```python
# BUILD: Write a function that safely converts any input to int.
# - If it's already an int, return it
# - If it's a float, truncate to int
# - If it's a string that looks like a number ("5" or "3.7"), convert it
# - If it's None or unconvertible, return 0

def safe_to_int(value):
    # TODO: write your solution
    if type(value) == int:
        return value 
    elif type(value) == float:
        return int(value)
    elif type(value) == str:
        try:
            return int(float(value))
        except ValueError:
            return 0
    elif type(value) == bool:
        return int(value)
    else:
        return 0

# Tests
assert safe_to_int(42) == 42
assert safe_to_int(3.7) == 3
assert safe_to_int("5") == 5
assert safe_to_int("3.7") == 3
assert safe_to_int(None) == 0
assert safe_to_int("hello") == 0
assert safe_to_int(True) == 1
print("All tests passed ✓")
```

## Section 5: Visualize & Explore

Run these cells to build intuition about how Python handles types.

```python
# EXPLORE: Float precision — see it in action

print("=== Float Precision Surprises ===")
print(f"0.1 + 0.2        = {0.1 + 0.2}")
print(f"0.1 + 0.2 == 0.3 = {0.1 + 0.2 == 0.3}")
print(f"0.1 + 0.2 - 0.3  = {0.1 + 0.2 - 0.3}")
print()
print(f"0.3 + 0.3 + 0.3      = {0.3 + 0.3 + 0.3}")
print(f"0.3 + 0.3 + 0.3 == 0.9 = {0.3 + 0.3 + 0.3 == 0.9}")
print()
print("Lesson: NEVER compare floats with ==. Use abs(a - b) < threshold.")
```

```python
# EXPLORE: Python has no integer overflow!

print("=== Big Numbers ===")
print(f"2 ** 10   = {2 ** 10}")
print(f"2 ** 100  = {2 ** 100}")
print(f"2 ** 1000 has {len(str(2 ** 1000))} digits!")
print()
print("Python handles arbitrarily large integers natively.")
print("Most other languages (C, Java) would overflow.")
```
