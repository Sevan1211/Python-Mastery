```python
# E2: Performance test — how much faster is set lookup vs list lookup?
import time

big_list = list(range(1_000_000))
big_set = set(big_list)

# Search for the LAST element (worst case for list)
target = 999_999

start = time.perf_counter()
for _ in range(100):
    _ = target in big_list
list_time = time.perf_counter() - start

start = time.perf_counter()
for _ in range(100):
    _ = target in big_set
set_time = time.perf_counter() - start

print(f"List: {list_time:.4f}s")
print(f"Set:  {set_time:.6f}s")
print(f"Set is {list_time / max(set_time, 1e-9):.0f}x faster")

# Your observation: ___
```
