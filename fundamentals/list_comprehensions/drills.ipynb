## Section 5: Explore & Discover

Open-ended cells. Run them, change them, see what happens.

```
# Explore 1: How does the order of nested for clauses affect the output?
# Predict, then run, then SWAP the two for clauses.
print([(x, y) for x in [1, 2, 3] for y in ['a', 'b']])
# Now swap:
print([(x, y) for y in ['a', 'b'] for x in [1, 2, 3]])
```

```
# Explore 2: Generator vs list — measure memory.
import sys
lst = [n for n in range(10000)]
gen = (n for n in range(10000))
print("list bytes:", sys.getsizeof(lst))
print("generator bytes:", sys.getsizeof(gen))
```

```
# Explore 3: What happens with empty inputs?
print([n * 2 for n in []])
print({n: n for n in []})
print([x for row in [] for x in row])
```

```
# Explore 4: Walrus operator — compute once, use twice.
def expensive(x):
    print(f"  computing {x}...")
    return x * x

# Without walrus — runs expensive() twice per item that passes the filter
print("Without walrus:")
result = [expensive(x) for x in range(5) if expensive(x) > 5]

print("\nWith walrus:")
result = [y for x in range(5) if (y := expensive(x)) > 5]
```

```
# Explore 5: Short-circuit with any() / all() on a generator.
def check(x):
    print(f"  checking {x}")
    return x > 3

print("any:")
print(any(check(x) for x in range(10)))
print("\nall:")
print(all(check(x) for x in range(10)))
```

```
# Explore 6: Compare timing — comprehension vs map() vs for-loop.
import timeit

t_comp = timeit.timeit("[x*x for x in range(1000)]", number=10000)
t_map  = timeit.timeit("list(map(lambda x: x*x, range(1000)))", number=10000)
t_loop = timeit.timeit('''
out = []
for x in range(1000):
    out.append(x*x)
''', number=10000)
print(f"comp: {t_comp:.3f}s")
print(f"map : {t_map:.3f}s")
print(f"loop: {t_loop:.3f}s")
```
