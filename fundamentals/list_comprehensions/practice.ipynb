# List Comprehensions — Practice Drills

**30 drills across 6 difficulty levels.** Each problem is solvable with a list, set, dict, or generator comprehension (sometimes combined with `zip`, `enumerate`, or `sorted`). Most are direct interview patterns or scaled-down versions of what you'll see in coding rounds.

For each drill: write the function body, run the cell, the asserts will tell you if it's correct.

**Levels**
- Level 1: Basic transforms & filters (Drills 1–5)
- Level 2: Conditionals, enumerate, zip (Drills 6–10)
- Level 3: Set & dict comprehensions (Drills 11–15)
- Level 4: 2D / nested / matrix patterns (Drills 16–20)
- Level 5: Real interview patterns — strings, records, frequency (Drills 21–25)
- Level 6: Harder / multi-step (Drills 26–30)

**Rule:** prefer a single comprehension where reasonable. If a problem clearly needs a helper line (e.g. building a frequency dict first), that's fine — but the *core* transformation should be a comprehension.
