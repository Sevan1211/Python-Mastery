## Section 5: Explore & Discover

```
# Explore 1: Stability — confirm Python's sort is stable.
data = [("A", 2), ("B", 1), ("C", 2), ("D", 1), ("E", 2)]
print(sorted(data, key=lambda x: x[1]))
```

```
# Explore 2: Two-pass sort for mixed asc/desc on non-numeric secondary.
# Goal: sort by (department ASC, hire_date DESC).
employees = [
    ("Alice",  "Eng",   "2020-01-01"),
    ("Bob",    "HR",    "2019-05-12"),
    ("Carol",  "Eng",   "2022-03-15"),
    ("Dave",   "HR",    "2021-07-30"),
    ("Eve",    "Eng",   "2018-11-20"),
]
# Two-pass: secondary first, primary second
step1 = sorted(employees, key=lambda e: e[2], reverse=True)
final = sorted(step1, key=lambda e: e[1])
for row in final:
    print(row)
```

```
# Explore 3: Compare timing — lambda vs itemgetter on a large list.
import timeit
from operator import itemgetter

data = [(i % 100, i) for i in range(100_000)]
t1 = timeit.timeit(lambda: sorted(data, key=lambda x: x[0]), number=20)
t2 = timeit.timeit(lambda: sorted(data, key=itemgetter(0)), number=20)
print(f"lambda:     {t1:.3f}s")
print(f"itemgetter: {t2:.3f}s")
```

```
# Explore 4: heapq.nlargest vs sorted+slice for top-K.
import timeit, heapq, random
random.seed(0)
big = [random.randint(0, 10**6) for _ in range(200_000)]
k = 5

t_sort = timeit.timeit(lambda: sorted(big, reverse=True)[:k], number=20)
t_heap = timeit.timeit(lambda: heapq.nlargest(k, big), number=20)
print(f"sorted+slice: {t_sort:.3f}s")
print(f"nlargest:     {t_heap:.3f}s")
```

```
# Explore 5: Sorting heterogeneous types — what blows up, what doesn't?
try:
    print(sorted([1, "a", 2]))
except TypeError as e:
    print("TypeError:", e)

# Workaround: provide a key that produces comparable values.
mixed = [3, "1", 2, "10"]
print(sorted(mixed, key=str))   # all compared as strings
```

```
# Explore 6: Custom alphabet.
alphabet = "zyxwvutsrqponmlkjihgfedcba"  # reversed alphabet
order = {ch: i for i, ch in enumerate(alphabet)}
words = ["apple", "banana", "cherry"]
print(sorted(words, key=lambda w: [order[c] for c in w]))
```
