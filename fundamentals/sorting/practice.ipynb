# Sorting — Practice Drills

**30 drills across 6 levels.** Heavy emphasis on the patterns that appear in interviews:
the `key` parameter, multi-field sorts, sorting records, custom comparators, and the
classic sort-then-scan algorithms (intervals, anagrams, top-K, frequency).

Each drill: write the function body, run the cell, asserts will tell you if it's right.

**Levels**
- L1: `sorted` vs `.sort()`, basic key (1–5)
- L2: Multi-field & mixed asc/desc (6–10)
- L3: Records, dicts, objects (11–15)
- L4: Frequency & top-K (16–20)
- L5: Classic interview patterns (21–25)
- L6: Harder — custom comparator, intervals, hard ordering (26–30)
