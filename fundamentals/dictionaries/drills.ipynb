```python
# E2: Explore dict view objects — are they "live"?
# What happens to keys/values/items views when you modify the dict?

d = {"a": 1, "b": 2}
keys = d.keys()
vals = d.values()
items = d.items()

print("Before:", keys, vals, items)

d["c"] = 3
print("After adding 'c':", keys, vals, items)

del d["a"]
print("After deleting 'a':", keys, vals, items)

# What does this tell you about dict views?
```

```python
# E3: Explore insertion order preservation.
# Create a dict, delete some keys, re-add them. Where do they end up?

d = {"first": 1, "second": 2, "third": 3, "fourth": 4}
print("Original:", list(d.keys()))

del d["second"]
print("After del 'second':", list(d.keys()))

d["second"] = 2
print("After re-adding 'second':", list(d.keys()))

# Does updating an existing key change its position?
d["first"] = 999
print("After updating 'first':", list(d.keys()))
```

```python
# E4: Explore the True/1 and False/0 key collision.
# Build a dict step by step and observe what happens.

d = {}
d[True] = "i am True"
print("After True:", d)

d[1] = "i am 1"
print("After 1:", d)

d[1.0] = "i am 1.0"
print("After 1.0:", d)

# How many keys are in d? Why?
print("Length:", len(d))
print("True == 1:", True == 1)
print("1 == 1.0:", 1 == 1.0)
print("hash(True):", hash(True), "hash(1):", hash(1), "hash(1.0):", hash(1.0))
```

```python
# E5: Explore performance — is dict lookup really O(1)?
# Time how long it takes to check membership in a dict vs a list.

import time

# Build a large dict and a large list
n = 1_000_000
big_dict = {i: True for i in range(n)}
big_list = list(range(n))

target = n - 1  # worst case for list (last element)

# Time dict lookup
start = time.time()
for _ in range(10000):
    _ = target in big_dict
dict_time = time.time() - start

# Time list lookup
start = time.time()
for _ in range(10000):
    _ = target in big_list
list_time = time.time() - start

print(f"Dict lookup (10k times): {dict_time:.4f}s")
print(f"List lookup (10k times): {list_time:.4f}s")
print(f"Dict is {list_time/dict_time:.0f}x faster")
```
