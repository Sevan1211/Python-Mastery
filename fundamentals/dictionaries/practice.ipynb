## Level 1 ⭐ — Basics

```python
# Drill 1: Return the value for a given key, or "NOT FOUND" if the key doesn't exist.
# {"a": 1, "b": 2}, "a" → 1   |   {"a": 1}, "z" → "NOT FOUND"

def safe_get(d: dict, key: str):
    # TODO
    return d.get(key, "NOT FOUND")

assert safe_get({"a": 1, "b": 2}, "a") == 1
assert safe_get({"a": 1, "b": 2}, "b") == 2
assert safe_get({"a": 1}, "z") == "NOT FOUND"
assert safe_get({}, "x") == "NOT FOUND"
print("Drill 1 passed ✓")
```

```python
# Drill 2: Return the number of key-value pairs in a dict.
# {"a": 1, "b": 2, "c": 3} → 3   |   {} → 0

def dict_size(d: dict) -> int:
    # TODO
    count = 0
    for i in d:
        count += 1
    
    return count
assert dict_size({"a": 1, "b": 2, "c": 3}) == 3
assert dict_size({}) == 0
assert dict_size({"x": None}) == 1
print("Drill 2 passed ✓")
```

```python
# Drill 3: Return True if a given key exists in the dict, False otherwise.
# {"name": "Alice"}, "name" → True   |   {"name": "Alice"}, "age" → False

def has_key(d: dict, key) -> bool:
    # TODO
    if key in d:
        return True
    return False

assert has_key({"name": "Alice", "age": 30}, "name") == True
assert has_key({"name": "Alice"}, "age") == False
assert has_key({}, "x") == False
assert has_key({0: "zero"}, 0) == True
print("Drill 3 passed ✓")
```

```python
# Drill 4: Create a dict from two lists (keys and values).
# ["a", "b"], [1, 2] → {"a": 1, "b": 2}

def lists_to_dict(keys: list, values: list) -> dict:
    # TODO

    return dict(zip(keys, values))

assert lists_to_dict(["a", "b"], [1, 2]) == {"a": 1, "b": 2}
assert lists_to_dict([], []) == {}
assert lists_to_dict(["x"], [10]) == {"x": 10}
assert lists_to_dict(["name", "age"], ["Alice", 30]) == {"name": "Alice", "age": 30}
print("Drill 4 passed ✓")
```

## Level 2 ⭐⭐ — Methods & Iteration

```python
# Drill 5: Return a list of all keys that have a value greater than a threshold.
# {"a": 10, "b": 5, "c": 20}, 8 → ["a", "c"]

def keys_above(d: dict, threshold: int) -> list:
    # TODO
    results = []
    for key, value in d.items():
        if value > threshold:
            results.append(key)
    
    return results

assert sorted(keys_above({"a": 10, "b": 5, "c": 20}, 8)) == ["a", "c"]
assert keys_above({"a": 1, "b": 2}, 10) == []
assert keys_above({}, 0) == []
assert sorted(keys_above({"x": 5, "y": 5, "z": 5}, 4)) == ["x", "y", "z"]
print("Drill 5 passed ✓")
```

```python
# Drill 6: Return a new dict with all values doubled.
# {"a": 5, "b": 10} → {"a": 10, "b": 20}

def double_values(d: dict) -> dict:
    # TODO
    results = {}

    for key, value in d.items():
        results[key] = value * 2

    return results
assert double_values({"a": 5, "b": 10}) == {"a": 10, "b": 20}
assert double_values({"x": 0}) == {"x": 0}
assert double_values({}) == {}
assert double_values({"a": -3, "b": 7}) == {"a": -6, "b": 14}
print("Drill 6 passed ✓")
```

```python
# Drill 7: Return a new dict with only the key-value pairs where value is a string.
# {"a": 1, "b": "hello", "c": True, "d": "world"} → {"b": "hello", "d": "world"}

def strings_only(d: dict) -> dict:
    # TODO
    results = {}

    for key, value in d.items():
        if isinstance(value,str):
            results[key] = value

    return results

assert strings_only({"a": 1, "b": "hello", "c": True, "d": "world"}) == {"b": "hello", "d": "world"}
assert strings_only({"x": 5}) == {}
assert strings_only({}) == {}
assert strings_only({"a": "yes", "b": "no"}) == {"a": "yes", "b": "no"}
print("Drill 7 passed ✓")
```

```python
# Drill 8: Return the sum of all values in a dict. If empty, return 0.
# {"a": 10, "b": 20, "c": 30} → 60

def sum_values(d: dict) -> int:
    # TODO
    return sum(d.values())

assert sum_values({"a": 10, "b": 20, "c": 30}) == 60
assert sum_values({}) == 0
assert sum_values({"x": -5, "y": 5}) == 0
assert sum_values({"only": 42}) == 42
print("Drill 8 passed ✓")
```

## Level 3 ⭐⭐⭐ — Multi-Step Transformations

```python
# Drill 9: Count character frequencies in a string.
# Return a dict mapping each character to its count.
# "aabbc" → {"a": 2, "b": 2, "c": 1}

def char_frequency(s: str) -> dict:
    # TODO

    result = {}

    for ch in s:
        result[ch] = result.get(ch, 0) + 1

    return result

assert char_frequency("aabbc") == {"a": 2, "b": 2, "c": 1}
assert char_frequency("") == {}
assert char_frequency("aaaa") == {"a": 4}
assert char_frequency("abc") == {"a": 1, "b": 1, "c": 1}
print("Drill 9 passed ✓")
```

```python
# Drill 10: Group words by their first letter.
# ["apple", "ant", "banana", "bat"] → {"a": ["apple", "ant"], "b": ["banana", "bat"]}

def group_by_first_letter(words: list) -> dict:
    # TODO
    result = {}

    for word in words:
        result.setdefault(word[0],[]).append(word)

    return result

assert group_by_first_letter(["apple", "ant", "banana", "bat"]) == {"a": ["apple", "ant"], "b": ["banana", "bat"]}
assert group_by_first_letter([]) == {}
assert group_by_first_letter(["hello"]) == {"h": ["hello"]}
assert group_by_first_letter(["x", "y", "z"]) == {"x": ["x"], "y": ["y"], "z": ["z"]}
print("Drill 10 passed ✓")
```

```python
# Drill 11: Given a dict of student scores, return the average score.
# Round to 2 decimal places. Return 0.0 for empty dict.
# {"Alice": 90, "Bob": 80, "Charlie": 70} → 80.0

def average_score(scores: dict) -> float:
    # TODO
    if not scores:
        return 0.0 
    return round(sum(scores.values())/len(scores.keys()), 2)

assert average_score({"Alice": 90, "Bob": 80, "Charlie": 70}) == 80.0
assert average_score({}) == 0.0
assert average_score({"A": 100}) == 100.0
assert average_score({"A": 33, "B": 33, "C": 34}) == 33.33
print("Drill 11 passed ✓")
```

```python
# Drill 12: Return a dict comprehension that maps each word to its length.
# ["hello", "world", "hi"] → {"hello": 5, "world": 5, "hi": 2}

def word_lengths(words: list) -> dict:
    # TODO
    return {word: len(word) for word in words }

assert word_lengths(["hello", "world", "hi"]) == {"hello": 5, "world": 5, "hi": 2}
assert word_lengths([]) == {}
assert word_lengths(["a"]) == {"a": 1}
assert word_lengths(["python", "is", "great"]) == {"python": 6, "is": 2, "great": 5}
print("Drill 12 passed ✓")
```

```python
# Drill 13: Two Sum — Return indices of two numbers that add up to target.
# Use a dict for O(n) solution. Return [] if no pair found.
# [2, 7, 11, 15], 9 → [0, 1]   |   [3, 2, 4], 6 → [1, 2]

def two_sum(nums: list, target: int) -> list:
    # TODO
    seen = {}

    for i, num in enumerate(nums):
        diff = target - num
        if diff in seen:
            return [seen[diff], i]
        
        seen[num] = i

    return []

assert two_sum([2, 7, 11, 15], 9) == [0, 1]
assert two_sum([3, 2, 4], 6) == [1, 2]
assert two_sum([3, 3], 6) == [0, 1]
assert two_sum([1, 2, 3], 10) == []
print("Drill 13 passed ✓")
```

```python
# Drill 14: Find the first non-repeating character in a string.
# Return the character, or "" if all repeat.
# "aabbc" → "c"   |   "aabb" → ""

def first_unique(s: str) -> str:
    # TODO
    repeating = {}

    for ch in s:
        repeating[ch] = repeating.get(ch, 0) + 1
    
    for ch in s:
        if repeating[ch] == 1:
            return ch
    
    return  ''

assert first_unique("aabbc") == "c"
assert first_unique("aabb") == ""
assert first_unique("leetcode") == "l"
assert first_unique("loveleetcode") == "v"
assert first_unique("") == ""
assert first_unique("z") == "z"
print("Drill 14 passed ✓")
```

```python
# Drill 15: Check if two strings are anagrams using frequency counting.
# "listen", "silent" → True   |   "hello", "world" → False

def is_anagram(s1: str, s2: str) -> bool:
    # TODO
    if len(s1) != len(s2):
        return False
    freq = {}

    for ch in s1:
        freq[ch] = freq.get(ch, 0) + 1
    
    for ch in s2:
        if ch not in freq:
            return False
        freq[ch] -= 1
    
    return all(count == 0 for count in freq.values())

assert is_anagram("listen", "silent") == True
assert is_anagram("hello", "world") == False
assert is_anagram("", "") == True
assert is_anagram("a", "a") == True
assert is_anagram("ab", "ba") == True
assert is_anagram("aab", "abb") == False
assert is_anagram("rat", "car") == False
print("Drill 15 passed ✓")
```

```python
# Drill 17: Check if a list contains any duplicate values.
# Use a dict or set approach — must be O(n).
# [1, 2, 3, 1] → True   |   [1, 2, 3, 4] → False

def contains_duplicate(nums: list) -> bool:
    # TODO
    return len(nums) != len(set(nums))

assert contains_duplicate([1, 2, 3, 1]) == True
assert contains_duplicate([1, 2, 3, 4]) == False
assert contains_duplicate([]) == False
assert contains_duplicate([1]) == False
assert contains_duplicate([1, 1]) == True
print("Drill 17 passed ✓")
```

```python
# Drill 18: Return the most frequent element in a list.
# If there's a tie, return any of the most frequent.
# [1, 2, 2, 3, 3, 3] → 3   |   ["a", "b", "a"] → "a"

def most_frequent(lst: list):
    # TODO
    freq = {}

    for val in lst:
        freq[val] = freq.get(val,0) + 1

    return max(freq, key = freq.get)

assert most_frequent([1, 2, 2, 3, 3, 3]) == 3
assert most_frequent(["a", "b", "a"]) == "a"
assert most_frequent([5]) == 5
assert most_frequent([1, 1, 2, 2, 3, 3, 3]) == 3
print("Drill 18 passed ✓")
```
