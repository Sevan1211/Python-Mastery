## 1. Defining Functions with `def`

A function is created with the `def` keyword, followed by a name, parentheses with optional parameters, and a colon. The body is indented.

```python
def function_name(parameters):
    """Optional docstring."""
    # body
    return value  # optional
```

Key rules:
- Function names should be `snake_case`
- A function must be **defined before** it's called
- A function that doesn't explicitly `return` returns `None`

````python
# Basic function — no parameters, no return
def greet():
    print("Hello, world!")

greet()  # Call the function
greet()  # Call it again — reusable!
````

````python
# Function with parameters
def greet_person(name):
    print(f"Hello, {name}!")

greet_person("Alice")
greet_person("Bob")
````

````python
# Function with return value
def add(a, b):
    return a + b

result = add(3, 5)
print(result)  # 8

# You can use the return value directly
print(add(10, 20))  # 30

# Return value can be used in expressions
total = add(1, 2) + add(3, 4)
print(total)  # 10
````

````python
# Functions without return explicitly return None
def say_hello():
    print("Hello")

result = say_hello()
print(f"Return value: {result}")   # None
print(f"Type: {type(result)}")     # <class 'NoneType'>
````

## 2. Parameters vs Arguments

These terms are often used interchangeably, but they're technically different:

- **Parameter:** The variable name in the function **definition** (the placeholder)
- **Argument:** The actual value passed when **calling** the function

```python
def greet(name):    # 'name' is a PARAMETER
    print(name)

greet("Alice")      # "Alice" is an ARGUMENT
```

````python
# Multiple parameters
def describe_pet(name, animal, age):
    print(f"{name} is a {age}-year-old {animal}")

# Positional arguments — order matters
describe_pet("Buddy", "dog", 5)

# What happens if you mix up the order?
describe_pet(5, "Buddy", "dog")  # Wrong — 5 is treated as name
````

## 3. Return Values — Multiple Returns and Early Returns

Functions can return any type, return multiple values (as a tuple), or return early from different branches.

````python
# Return multiple values (Python packs them into a tuple)
def min_max(nums):
    return min(nums), max(nums)

result = min_max([3, 1, 4, 1, 5, 9])
print(result)        # (1, 9)
print(type(result))  # <class 'tuple'>

# Unpack the tuple
lo, hi = min_max([3, 1, 4, 1, 5, 9])
print(f"Min: {lo}, Max: {hi}")  # Min: 1, Max: 9
````

````python
# Early return — guard clauses
# This pattern is cleaner than deeply nested if/else

def divide(a, b):
    if b == 0:
        return None  # Early return for edge case
    return a / b

print(divide(10, 3))  # 3.333...
print(divide(10, 0))  # None

# Compare with the nested version (worse):
def divide_nested(a, b):
    if b != 0:
        result = a / b
        return result
    else:
        return None
````

````python
# Return ends the function immediately
def first_negative(nums):
    for n in nums:
        if n < 0:
            return n  # Stops the entire function, not just the loop
    return None  # No negative found

print(first_negative([1, 2, -3, 4]))  # -3
print(first_negative([1, 2, 3]))      # None
````

## 4. Default Parameters

Parameters can have default values. If the caller doesn't provide that argument, the default is used.

**Rule:** Default parameters must come **after** non-default parameters.

````python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Alice")              # Uses default: Hello, Alice!
greet("Bob", "Hey")         # Override: Hey, Bob!
greet("Charlie", greeting="Howdy")  # Keyword argument
````

````python
# Multiple defaults
def create_user(name, role="viewer", active=True):
    return {"name": name, "role": role, "active": active}

print(create_user("Alice"))                          # All defaults
print(create_user("Bob", "admin"))                   # Override role
print(create_user("Charlie", active=False))          # Skip role, set active
print(create_user("Dana", "editor", False))          # Override both
````

````python
# DANGER: Mutable default argument — the #1 Python gotcha with functions

# BAD — the default list is shared across ALL calls!
def append_to_bad(item, lst=[]):
    lst.append(item)
    return lst

print(append_to_bad(1))  # [1]
print(append_to_bad(2))  # [1, 2] ← Surprise! Not [2]!
print(append_to_bad(3))  # [1, 2, 3] ← Keeps growing!

print("---")

# GOOD — use None as default, create new list inside
def append_to_good(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(append_to_good(1))  # [1]
print(append_to_good(2))  # [2] ← Correct!
print(append_to_good(3))  # [3] ← Correct!
````

## 5. Keyword Arguments

You can pass arguments by name instead of position. This makes function calls more readable, especially with many parameters.

````python
def send_email(to, subject, body, cc=None, urgent=False):
    print(f"To: {to}")
    print(f"Subject: {subject}")
    print(f"Body: {body}")
    if cc:
        print(f"CC: {cc}")
    if urgent:
        print("⚠️ URGENT")
    print()

# Positional
send_email("alice@example.com", "Meeting", "See you at 3pm")

# Keyword arguments — much clearer
send_email(
    to="bob@example.com",
    subject="Reminder",
    body="Don't forget the deadline",
    urgent=True
)
````

````python
# Rule: Positional arguments must come BEFORE keyword arguments

def f(a, b, c):
    print(a, b, c)

f(1, 2, 3)           # All positional ✓
f(1, c=3, b=2)       # Mix positional + keyword ✓
f(a=1, b=2, c=3)     # All keyword ✓

# f(a=1, 2, 3)       # SyntaxError! keyword before positional ✗
````

## 6. `*args` and `**kwargs` — Variable Arguments

### `*args` — Variable Positional Arguments

Collects extra positional arguments into a **tuple**.

### `**kwargs` — Variable Keyword Arguments

Collects extra keyword arguments into a **dictionary**.

````python
# *args — accept any number of positional arguments
def add_all(*args):
    print(f"args = {args}")       # It's a tuple
    print(f"type = {type(args)}") 
    return sum(args)

print(add_all(1, 2, 3))       # 6
print(add_all(10, 20))        # 30
print(add_all(5))             # 5
print(add_all())              # 0
````

````python
# **kwargs — accept any number of keyword arguments
def print_info(**kwargs):
    print(f"kwargs = {kwargs}")       # It's a dict
    print(f"type = {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

print_info(name="Alice", age=30, role="engineer")
````

````python
# Combining regular params, *args, **kwargs
# Order MUST be: regular → *args → **kwargs

def flexible(required, *args, **kwargs):
    print(f"required = {required}")
    print(f"args = {args}")
    print(f"kwargs = {kwargs}")

flexible("hello", 1, 2, 3, x=10, y=20)
# required = hello
# args = (1, 2, 3)
# kwargs = {'x': 10, 'y': 20}
````

````python
# Unpacking into function calls with * and **

def add(a, b, c):
    return a + b + c

# Unpack a list/tuple with *
nums = [1, 2, 3]
print(add(*nums))  # Same as add(1, 2, 3) → 6

# Unpack a dict with **
params = {"a": 10, "b": 20, "c": 30}
print(add(**params))  # Same as add(a=10, b=20, c=30) → 60
````

## 7. Keyword-Only and Positional-Only Parameters

Python 3 lets you enforce how arguments are passed:

- **Keyword-only:** Parameters after `*` must be passed by name
- **Positional-only:** Parameters before `/` must be passed by position (Python 3.8+)

````python
# Keyword-only parameters — everything after * must be keyword
def connect(host, port, *, timeout=30, retries=3):
    print(f"Connecting to {host}:{port} (timeout={timeout}, retries={retries})")

connect("localhost", 5432)                        # ✓
connect("localhost", 5432, timeout=10)             # ✓
# connect("localhost", 5432, 10)                   # TypeError! timeout is keyword-only
````

````python
# Positional-only parameters — everything before / must be positional
def power(base, exp, /):
    return base ** exp

print(power(2, 10))     # ✓ → 1024
# print(power(base=2, exp=10))  # TypeError! positional-only
````

````python
# Combined: positional-only / regular / * keyword-only
def func(a, b, /, c, d, *, e, f):
    print(a, b, c, d, e, f)

# a, b: positional only
# c, d: either 
# e, f: keyword only
func(1, 2, 3, d=4, e=5, f=6)  # ✓
````

## 8. Scope — LEGB Rule

When Python sees a variable name, it searches for it in this order:

1. **L**ocal — inside the current function
2. **E**nclosing — inside enclosing functions (closures)
3. **G**lobal — at the module level
4. **B**uilt-in — Python's built-in names (`len`, `print`, `range`, etc.)

First match wins.

````python
# Local scope
def func():
    x = 10  # local to func
    print(f"Inside func: x = {x}")

func()
# print(x)  # NameError — x doesn't exist outside func
````

````python
# Global scope
counter = 0  # global variable

def increment():
    global counter  # Declare we want to modify the global
    counter += 1

increment()
increment()
print(counter)  # 2

# NOTE: Avoid 'global' in production code — it makes functions impure
# and harder to test. Pass values as parameters and return results.
````

````python
# Enclosing scope (closures)
def outer():
    message = "hello"  # Enclosing scope
    
    def inner():
        print(message)  # Accesses enclosing variable
    
    inner()

outer()  # "hello"
````

````python
# LEGB demo — same name at different scopes
x = "global"

def outer():
    x = "enclosing"
    
    def inner():
        x = "local"
        print(f"inner: {x}")   # local
    
    inner()
    print(f"outer: {x}")       # enclosing

outer()
print(f"global: {x}")          # global
````

````python
# nonlocal — modify enclosing scope variable
def counter_factory():
    count = 0
    
    def increment():
        nonlocal count  # Modify the enclosing variable
        count += 1
        return count
    
    return increment

counter = counter_factory()
print(counter())  # 1
print(counter())  # 2
print(counter())  # 3
````

## 9. Docstrings

A docstring is a string literal as the **first statement** in a function. It documents what the function does and is accessible via `help()` and `.__doc__`.

````python
def binary_search(arr, target):
    """Search for target in a sorted array.
    
    Args:
        arr: A sorted list of comparable elements.
        target: The value to search for.
    
    Returns:
        The index of target if found, -1 otherwise.
    
    Examples:
        >>> binary_search([1, 3, 5, 7], 5)
        2
        >>> binary_search([1, 3, 5, 7], 4)
        -1
    """
    lo, hi = 0, len(arr) - 1
    while lo <= hi:
        mid = (lo + hi) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
    return -1

print(binary_search([1, 3, 5, 7, 9], 5))  # 2
print(binary_search.__doc__)  # View the docstring
help(binary_search)           # Formatted help
````

## 10. Type Hints (Annotations)

Type hints don't enforce types at runtime — they're **documentation** that tools like `mypy` and IDEs use for static checking. They make code much more readable.

````python
# Basic type hints
def add(a: int, b: int) -> int:
    return a + b

def greet(name: str, loud: bool = False) -> str:
    msg = f"Hello, {name}!"
    return msg.upper() if loud else msg

print(add(3, 5))              # 8
print(greet("Alice"))         # Hello, Alice!
print(greet("Bob", loud=True))  # HELLO, BOB!

# Type hints are NOT enforced — this still runs:
print(add("hello", " world"))  # "hello world" — no error!
````

````python
# Common type hints
from typing import Optional

def find_user(user_id: int) -> Optional[dict]:
    """Returns user dict or None if not found."""
    users = {1: {"name": "Alice"}, 2: {"name": "Bob"}}
    return users.get(user_id)

# Modern Python (3.10+) uses | instead of Optional
def find_user_modern(user_id: int) -> dict | None:
    users = {1: {"name": "Alice"}, 2: {"name": "Bob"}}
    return users.get(user_id)

print(find_user(1))   # {'name': 'Alice'}
print(find_user(99))  # None
````

````python
# Collection type hints

# Python 3.9+: use built-in types directly
def process_items(items: list[str]) -> dict[str, int]:
    return {item: len(item) for item in items}

print(process_items(["hello", "world", "hi"]))
# {'hello': 5, 'world': 5, 'hi': 2}

def combine(a: tuple[int, ...], b: tuple[int, ...]) -> list[int]:
    return sorted(list(a) + list(b))

print(combine((3, 1), (4, 2)))  # [1, 2, 3, 4]
````

## 11. First-Class Functions

In Python, functions are **first-class objects** — they can be:
- Assigned to variables
- Passed as arguments to other functions
- Returned from other functions
- Stored in data structures

This is one of Python's most powerful features.

````python
# Assign a function to a variable
def square(x):
    return x ** 2

f = square  # f now refers to the same function
print(f(5))        # 25
print(f is square) # True — same object
````

````python
# Pass a function as an argument
def apply_to_list(func, items):
    return [func(item) for item in items]

def double(x):
    return x * 2

def negate(x):
    return -x

nums = [1, 2, 3, 4, 5]
print(apply_to_list(double, nums))  # [2, 4, 6, 8, 10]
print(apply_to_list(negate, nums))  # [-1, -2, -3, -4, -5]
print(apply_to_list(str, nums))     # ['1', '2', '3', '4', '5']
````

````python
# Built-in functions that accept functions as arguments
nums = [3, -1, 4, -1, 5, -9]

# sorted() with a key function
print(sorted(nums))                   # [-9, -1, -1, 3, 4, 5]
print(sorted(nums, key=abs))           # [-1, -1, 3, 4, 5, -9]

# map() — apply function to every element
print(list(map(abs, nums)))            # [3, 1, 4, 1, 5, 9]

# filter() — keep elements where function returns True
print(list(filter(lambda x: x > 0, nums)))  # [3, 4, 5]
````

````python
# Functions in data structures
operations = {
    "add": lambda a, b: a + b,
    "sub": lambda a, b: a - b,
    "mul": lambda a, b: a * b,
}

op = "add"
print(operations[op](10, 3))  # 13

op = "mul"
print(operations[op](10, 3))  # 30
````

## 12. Lambda Functions

Small anonymous functions defined in a single expression.

Syntax: `lambda parameters: expression`

Lambdas can only contain a **single expression** (no statements, no assignments, no multiline).

````python
# Basic lambda
square = lambda x: x ** 2
print(square(5))  # 25

# Multiple parameters
add = lambda a, b: a + b
print(add(3, 7))  # 10

# Lambda with conditional
classify = lambda n: "even" if n % 2 == 0 else "odd"
print(classify(4))  # even
print(classify(7))  # odd
````

````python
# Most common use: as a key function in sorting
students = [
    {"name": "Alice", "grade": 88},
    {"name": "Bob", "grade": 72},
    {"name": "Charlie", "grade": 95},
]

# Sort by grade
by_grade = sorted(students, key=lambda s: s["grade"])
print([s["name"] for s in by_grade])  # ['Bob', 'Alice', 'Charlie']

# Sort by name length
by_name_len = sorted(students, key=lambda s: len(s["name"]))
print([s["name"] for s in by_name_len])  # ['Bob', 'Alice', 'Charlie']

# Sort strings case-insensitively
words = ["Banana", "apple", "Cherry"]
print(sorted(words))                         # ['Banana', 'Cherry', 'apple']
print(sorted(words, key=lambda w: w.lower())) # ['apple', 'Banana', 'Cherry']
````

````python
# When NOT to use lambda

# BAD — lambda assigned to a name (use def instead)
# square = lambda x: x ** 2  # PEP 8 says: use def

# GOOD
def square(x):
    return x ** 2

# Lambda is best for SHORT, throwaway functions used inline
# If the lambda is complex, use a named function for readability
````

## 13. Closures

A closure is a function that **remembers** the variables from its enclosing scope, even after the outer function has finished executing. This is how `nonlocal` works under the hood.

````python
# A closure that remembers the multiplier
def make_multiplier(factor):
    def multiply(x):
        return x * factor  # 'factor' comes from enclosing scope
    return multiply

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(5))   # 10
print(triple(5))   # 15
print(double(10))  # 20

# Each closure has its OWN copy of 'factor'
print(double.__closure__[0].cell_contents)  # 2
print(triple.__closure__[0].cell_contents)  # 3
````

````python
# Closure as a lightweight alternative to a class
def make_counter(start=0):
    count = start
    
    def counter():
        nonlocal count
        count += 1
        return count
    
    return counter

c1 = make_counter()
c2 = make_counter(100)

print(c1(), c1(), c1())  # 1 2 3
print(c2(), c2())        # 101 102
````

````python
# Common closure gotcha — late binding in loops

# BAD — all functions share the SAME variable i
funcs_bad = []
for i in range(3):
    funcs_bad.append(lambda: i)

print([f() for f in funcs_bad])  # [2, 2, 2] ← All return 2!

# FIX — capture the value with a default argument
funcs_good = []
for i in range(3):
    funcs_good.append(lambda i=i: i)  # Default captures current value

print([f() for f in funcs_good])  # [0, 1, 2] ← Correct!
````

## 14. Decorators

A decorator is a function that **wraps** another function, adding behavior before and/or after it. Decorators use the `@` syntax.

Pattern:
```python
@decorator
def func():
    ...

# Equivalent to:
func = decorator(func)
```

````python
# A simple decorator that logs function calls
import functools

def log_calls(func):
    @functools.wraps(func)  # Preserves original function's metadata
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}({args}, {kwargs})")
        result = func(*args, **kwargs)
        print(f"  → returned {result}")
        return result
    return wrapper

@log_calls
def add(a, b):
    return a + b

@log_calls
def greet(name):
    return f"Hello, {name}!"

add(3, 5)
greet("Alice")
````

````python
# Practical decorator: timing
import time
import functools

def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f"{func.__name__} took {elapsed:.4f}s")
        return result
    return wrapper

@timer
def slow_sum(n):
    return sum(range(n))

slow_sum(10_000_000)
````

````python
# Practical decorator: memoization (caching)
# Python has a built-in one: functools.lru_cache

from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n):
    if n < 2:
        return n
    return fib(n - 1) + fib(n - 2)

print(fib(50))  # 12586269025 — instant with caching
print(fib.cache_info())  # Shows cache hits/misses
````

## 15. Recursion Basics

A function that calls **itself**. Every recursive function needs:
1. **Base case** — when to stop
2. **Recursive case** — how to reduce toward the base case

Recursion gets its own full module later — this is just the fundamentals.

````python
# Factorial: n! = n * (n-1) * ... * 1
def factorial(n):
    if n <= 1:       # Base case
        return 1
    return n * factorial(n - 1)  # Recursive case

print(factorial(5))   # 120
print(factorial(0))   # 1
print(factorial(10))  # 3628800
````

````python
# Sum of a list using recursion
def recursive_sum(nums):
    if not nums:         # Base case: empty list
        return 0
    return nums[0] + recursive_sum(nums[1:])  # First + rest

print(recursive_sum([1, 2, 3, 4, 5]))  # 15
print(recursive_sum([]))                 # 0
````

````python
# Python's recursion limit
import sys
print(f"Default recursion limit: {sys.getrecursionlimit()}")

# For interview problems, recursion depth > 1000 usually means
# you should use an iterative approach or increase the limit.
````

## 16. Common Function Patterns

These patterns appear constantly in interviews and production code.

### 16a. Predicate Functions

Functions that return `True`/`False`. Used for filtering, validation, and conditions.

````python
def is_even(n):
    return n % 2 == 0

def is_positive(n):
    return n > 0

nums = [-3, -1, 0, 2, 4, 7]
print(list(filter(is_even, nums)))      # [0, 2, 4]
print(list(filter(is_positive, nums)))  # [2, 4, 7]

# all() and any() with predicates
print(all(is_positive(n) for n in [1, 2, 3]))   # True
print(any(is_even(n) for n in [1, 3, 5]))        # False
````

### 16b. Transformation / Map Functions

````python
def celsius_to_fahrenheit(c):
    return c * 9 / 5 + 32

temps_c = [0, 20, 37, 100]
temps_f = list(map(celsius_to_fahrenheit, temps_c))
print(temps_f)  # [32.0, 68.0, 98.6, 212.0]

# Same with list comprehension (usually preferred)
temps_f2 = [celsius_to_fahrenheit(c) for c in temps_c]
print(temps_f2)
````

### 16c. Helper Functions

````python
# Break complex logic into small, testable helpers

def is_valid_email(email):
    """Basic email validation."""
    return "@" in email and "." in email.split("@")[-1]

def clean_email(email):
    """Normalize an email address."""
    return email.strip().lower()

def process_emails(raw_emails):
    """Clean and validate a list of emails."""
    result = []
    for email in raw_emails:
        cleaned = clean_email(email)
        if is_valid_email(cleaned):
            result.append(cleaned)
    return result

emails = ["  Alice@Example.COM  ", "invalid", "bob@test.org", ""]
print(process_emails(emails))  # ['alice@example.com', 'bob@test.org']
````

### 16d. Factory Functions

````python
# Return different functions based on input
def make_validator(min_val, max_val):
    """Create a validation function for a range."""
    def validate(value):
        return min_val <= value <= max_val
    return validate

is_valid_age = make_validator(0, 150)
is_valid_score = make_validator(0, 100)

print(is_valid_age(25))     # True
print(is_valid_age(200))    # False
print(is_valid_score(85))   # True
print(is_valid_score(-5))   # False
````

## 17. Gotchas and Anti-Patterns

````python
# GOTCHA 1: Mutable default arguments (revisited — it's THAT important)

# BAD
def add_item_bad(item, lst=[]):
    lst.append(item)
    return lst

# GOOD
def add_item_good(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(add_item_good(1))  # [1]
print(add_item_good(2))  # [2] — correct!
````

````python
# GOTCHA 2: Forgetting to return a value

# BAD — prints but doesn't return
def add_bad(a, b):
    print(a + b)  # Output goes to console, returns None

result = add_bad(3, 5)  # Prints 8
print(f"result = {result}")  # result = None

# GOOD
def add_good(a, b):
    return a + b
````

````python
# GOTCHA 3: Shadowing built-in names

# BAD — don't use built-in names as parameter names!
# def process(list, dict):  # Shadows built-in list and dict
#     ...

# GOOD — use descriptive names
def process(items, config):
    print(f"Processing {len(items)} items with config: {config}")

# Other commonly shadowed names: id, type, input, str, int, max, min, sum
# If you accidentally shadow one: restart your Python session or del the name
````

````python
# GOTCHA 4: Modifying a mutable argument

def remove_negatives(nums):
    """Removes negatives IN PLACE — modifies the original list!"""
    i = 0
    while i < len(nums):
        if nums[i] < 0:
            nums.pop(i)
        else:
            i += 1

original = [1, -2, 3, -4, 5]
remove_negatives(original)
print(original)  # [1, 3, 5] — original was modified!

# If you don't want to modify the original, work on a copy:
def remove_negatives_safe(nums):
    return [n for n in nums if n >= 0]
````

````python
# GOTCHA 5: Global state instead of parameters

# BAD — depends on global state, hard to test
multiplier = 3
def multiply_bad(x):
    return x * multiplier

# GOOD — pure function, all inputs are parameters
def multiply_good(x, multiplier):
    return x * multiplier

# Pure functions are:
# - Easier to test
# - Easier to reason about
# - Safer in concurrent code
# - More reusable
````

## 18. Functions as Interview Tools

In technical interviews, functions are not just about code — they communicate your **thought process**.

````python
# Interview best practices:

# 1. Name functions clearly
def find_max_profit(prices):  # ✓ Clear intent
    pass

# def solve(p):  # ✗ What does this do?
#     pass

# 2. Handle edge cases first (guard clauses)
def find_max_profit(prices):
    if not prices or len(prices) < 2:
        return 0
    # Main logic here...
    min_price = prices[0]
    max_profit = 0
    for price in prices[1:]:
        max_profit = max(max_profit, price - min_price)
        min_price = min(min_price, price)
    return max_profit

print(find_max_profit([7, 1, 5, 3, 6, 4]))  # 5
print(find_max_profit([7, 6, 4, 3, 1]))      # 0
print(find_max_profit([]))                     # 0
````

````python
# 3. Extract helper functions for complex problems
def is_valid_parentheses(s):
    """Check if a string has valid matched parentheses."""
    stack = []
    matching = {')': '(', ']': '[', '}': '{'}
    
    for char in s:
        if char in '([{':
            stack.append(char)
        elif char in ')]}':
            if not stack or stack[-1] != matching[char]:
                return False
            stack.pop()
    
    return len(stack) == 0

print(is_valid_parentheses("()"))     # True
print(is_valid_parentheses("()[]{}")) # True
print(is_valid_parentheses("(]"))     # False
print(is_valid_parentheses("([)]"))   # False
print(is_valid_parentheses("{[]}"))   # True
````

## Summary

| Concept | Key Takeaway |
|---------|--------------|
| `def` | Define reusable, named blocks of code |
| Parameters vs arguments | Parameters are in the definition; arguments are in the call |
| `return` | Sends a value back; no return → `None` |
| Default params | Must come after non-default; **never use mutable defaults** |
| `*args`, `**kwargs` | Variable positional (tuple) and keyword (dict) arguments |
| Keyword-only (`*`) | After `*`, arguments must be passed by name |
| Positional-only (`/`) | Before `/`, arguments must be passed by position |
| LEGB scope | Local → Enclosing → Global → Built-in |
| `global` / `nonlocal` | Modify outer scope variables (use sparingly) |
| Type hints | Documentation for tools/IDEs; not enforced at runtime |
| First-class functions | Functions can be assigned, passed, returned, stored |
| Lambda | Anonymous one-expression functions; best for inline use |
| Closures | Inner functions remember enclosing scope variables |
| Decorators | `@decorator` wraps a function with extra behavior |
| Recursion | Function calls itself; needs base case + recursive case |
| Pure functions | No side effects, all inputs via parameters → easier to test |
