```python
# Explore 4: How does Python handle recursive depth?
import sys

def countdown(n):
    if n == 0:
        return "done"
    return countdown(n - 1)

print(f"Recursion limit: {sys.getrecursionlimit()}")
print(countdown(100))    # Works
print(countdown(500))    # Works

# Try: countdown(10000) — what happens?
# (Don't do it in production code!)

# Question: What's the default recursion limit? Can you change it?

# Your explanation:
```

```python
# Explore 5: Can you stack decorators?
def bold(func):
    def wrapper(*args, **kwargs):
        return f"<b>{func(*args, **kwargs)}</b>"
    return wrapper

def italic(func):
    def wrapper(*args, **kwargs):
        return f"<i>{func(*args, **kwargs)}</i>"
    return wrapper

@bold
@italic
def greet(name):
    return f"Hello, {name}"

print(greet("Alice"))

# Question: What order are decorators applied? Which is "inner" vs "outer"?

# Your explanation:
```
