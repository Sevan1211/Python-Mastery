## 1. What Is a List?

A list is an **ordered, mutable, heterogeneous** collection of items.

- **Ordered**: Items have positions (indices). Order is preserved.
- **Mutable**: You can change, add, or remove items after creation.
- **Heterogeneous**: A single list can hold different types (`int`, `str`, `bool`, other lists, etc.).

Lists are implemented as **dynamic arrays** under the hood — they auto-resize as you add elements.

```python
# A list is defined with square brackets
fruits = ["apple", "banana", "cherry"]
print(type(fruits))   # <class 'list'>
print(fruits)         # ['apple', 'banana', 'cherry']
print(len(fruits))    # 3
```

## 2. Creating Lists

There are many ways to create a list. Know all of them — they appear in interviews and production code.

```python
# --- Literal syntax (most common) ---
nums = [1, 2, 3, 4, 5]
print("Literal:", nums)

# --- Empty list ---
empty1 = []
empty2 = list()
print("Empty:", empty1, empty2)

# --- From a string (each character becomes an element) ---
chars = list("hello")
print("From string:", chars)  # ['h', 'e', 'l', 'l', 'o']

# --- From range ---
r = list(range(5))
print("From range:", r)  # [0, 1, 2, 3, 4]

r2 = list(range(2, 10, 3))
print("Range with step:", r2)  # [2, 5, 8]

# --- From split ---
words = "hello world foo".split()
print("From split:", words)  # ['hello', 'world', 'foo']

# --- Repetition ---
zeros = [0] * 5
print("Repetition:", zeros)  # [0, 0, 0, 0, 0]

# --- Mixed types (heterogeneous) ---
mixed = [1, "two", 3.0, True, None, [5, 6]]
print("Mixed:", mixed)
```

## 3. Indexing

Lists use **zero-based indexing**, exactly like strings.

- Positive indices count from the left: `0, 1, 2, ...`
- Negative indices count from the right: `-1, -2, -3, ...`

```python
colors = ["red", "green", "blue", "yellow", "purple"]

# Positive indexing
print(colors[0])    # red
print(colors[2])    # blue
print(colors[4])    # purple

# Negative indexing
print(colors[-1])   # purple (last)
print(colors[-2])   # yellow (second to last)
print(colors[-5])   # red (first — same as [0])
```

```python
# IndexError — accessing beyond the list
try:
    print(colors[10])  # IndexError!
except IndexError as e:
    print(f"IndexError: {e}")  # list index out of range

# Key difference from slicing: indexing raises errors, slicing doesn't
```

## 4. Slicing

Slicing works identically to strings: `list[start:stop:step]`

- `start`: inclusive (default 0)
- `stop`: exclusive (default len)
- `step`: step size (default 1)

**Slicing always returns a new list** (a shallow copy).

```python
nums = [10, 20, 30, 40, 50, 60, 70, 80]

# Basic slices
print(nums[1:4])     # [20, 30, 40]  — indices 1, 2, 3
print(nums[:3])      # [10, 20, 30]  — from start
print(nums[5:])      # [60, 70, 80]  — to end
print(nums[:])       # full copy

# With step
print(nums[::2])     # [10, 30, 50, 70]  — every other
print(nums[1::2])    # [20, 40, 60, 80]  — every other starting at 1

# Negative step (reverse)
print(nums[::-1])    # [80, 70, 60, 50, 40, 30, 20, 10]
print(nums[5:1:-1])  # [60, 50, 40, 30]  — reverse from index 5 to 2

# Out-of-bounds slicing is safe (no error)
print(nums[100:200]) # []  — no error!
```

```python
# IMPORTANT: Slice assignment (mutates the original list)
data = [1, 2, 3, 4, 5]

# Replace a range
data[1:3] = [20, 30]
print(data)  # [1, 20, 30, 4, 5]

# Insert via slice (replace nothing)
data[2:2] = [99, 98]
print(data)  # [1, 20, 99, 98, 30, 4, 5]

# Delete via slice (replace with nothing)
data[2:4] = []
print(data)  # [1, 20, 30, 4, 5]
```

## 5. Mutability — The Big Difference from Strings

Strings are **immutable** — you can't change a character in place.  
Lists are **mutable** — you CAN change elements in place.

This is the single most important conceptual shift from strings to lists.

```python
# Strings: IMMUTABLE
s = "hello"
try:
    s[0] = "H"  # TypeError!
except TypeError as e:
    print(f"String error: {e}")

# Lists: MUTABLE
lst = ["h", "e", "l", "l", "o"]
lst[0] = "H"  # Works!
print(lst)  # ['H', 'e', 'l', 'l', 'o']
```

```python
# Mutability means in-place modification
# Many list methods return None because they modify the list directly

nums = [3, 1, 2]
result = nums.sort()  # Sorts IN PLACE
print(nums)           # [1, 2, 3]
print(result)         # None  <-- THIS CATCHES EVERYONE

# Compare with sorted() which returns a NEW list
nums2 = [3, 1, 2]
result2 = sorted(nums2)
print(nums2)           # [3, 1, 2]  — unchanged
print(result2)         # [1, 2, 3]  — new list
```

```python
# Aliasing — two variables pointing to the SAME list
a = [1, 2, 3]
print(a)
b = a           # b is NOT a copy — it's the SAME list
b.append(4)
print(a)        # [1, 2, 3, 4]  — a changed too!
print(a is b)   # True — same object in memory

# To make a copy, you need to be explicit:
c = a[:]        # Shallow copy via slice
d = a.copy()    # Shallow copy via method
e = list(a)     # Shallow copy via constructor

c.append(999)
print(a)        # [1, 2, 3, 4]  — unchanged
print(c)        # [1, 2, 3, 4, 999]
```

## 6. List Operators

| Operator | Meaning | Example |
|----------|---------|--------|
| `+` | Concatenation | `[1,2] + [3,4]` → `[1,2,3,4]` |
| `*` | Repetition | `[0] * 3` → `[0,0,0]` |
| `in` | Membership | `3 in [1,2,3]` → `True` |
| `not in` | Non-membership | `5 not in [1,2,3]` → `True` |
| `==` | Equality (element-wise) | `[1,2] == [1,2]` → `True` |
| `is` | Identity (same object) | `a is b` → depends |

```python
# Concatenation — creates a NEW list
a = [1, 2]
b = [3, 4]
c = a + b
print(c)  # [1, 2, 3, 4]
print(a)  # [1, 2]  — unchanged

# Repetition
print(["ha"] * 3)  # ['ha', 'ha', 'ha']

# Membership
print(3 in [1, 2, 3])       # True
print("x" not in ["a", "b"]) # True

# Equality vs Identity
x = [1, 2, 3]
y = [1, 2, 3]
print(x == y)   # True  — same content
print(x is y)   # False — different objects
```

```python
# GOTCHA: Repetition with mutable nested objects
# This is a classic bug!

grid = [[0] * 3] * 3  # Looks like a 3x3 grid
print(grid)  # [[0, 0, 0], [0, 0, 0], [0, 0, 0]]

grid[0][0] = 1
print(grid)  # [[1, 0, 0], [1, 0, 0], [1, 0, 0]]  — ALL rows changed!

# Why? All 3 inner lists are the SAME object.
print(grid[0] is grid[1])  # True

# CORRECT way to make a 2D grid:
grid2 = [[0] * 3 for _ in range(3)]
grid2[0][0] = 1
print(grid2)  # [[1, 0, 0], [0, 0, 0], [0, 0, 0]]  — only row 0 changed
```

## 7. Common List Methods — Overview

| Method | Returns | Mutates? | Description |
|--------|---------|----------|-------------|
| `append(x)` | `None` | Yes | Add x to end |
| `insert(i, x)` | `None` | Yes | Insert x at index i |
| `extend(iterable)` | `None` | Yes | Add all items from iterable |
| `remove(x)` | `None` | Yes | Remove first occurrence of x |
| `pop(i)` | removed item | Yes | Remove & return item at index i |
| `clear()` | `None` | Yes | Remove all items |
| `index(x)` | int | No | Index of first x |
| `count(x)` | int | No | # of occurrences of x |

| `copy()` | new list | No | Shallow copy |

**The #1 beginner mistake**: methods that mutate return `None`, so `x = lst.sort()` sets x to None.

## 8. Adding Elements — append, insert, extend

Three ways to add items. Know the differences — they behave differently and have different time complexities.

```python
# --- append(x) — O(1) amortized ---
# Adds ONE item to the END
fruits = ["apple", "banana"]
fruits.append("cherry")
print(fruits)  # ['apple', 'banana', 'cherry']

# append adds the WHOLE object as a single element
fruits.append(["date", "elderberry"])
print(fruits)  # ['apple', 'banana', 'cherry', ['date', 'elderberry']]
# ^ It's nested! Not flat!
```

```python
# --- insert(index, x) — O(n) ---
# Inserts at a specific position, shifting everything after it
nums = [1, 2, 3, 4]
nums.insert(0, 0)     # Insert 0 at the beginning
print(nums)           # [0, 1, 2, 3, 4]

nums.insert(2, 99)    # Insert 99 at index 2
print(nums)           # [0, 1, 99, 2, 3, 4]

nums.insert(100, 999) # Index beyond length → appends
print(nums)           # [0, 1, 99, 2, 3, 4, 999]
```

```python
# --- extend(iterable) — O(k) where k = len(iterable) ---
# Adds each item from the iterable individually (flattens one level)
a = [1, 2, 3]
a.extend([4, 5, 6])
print(a)  # [1, 2, 3, 4, 5, 6]  — flat!

# extend vs append comparison
b = [1, 2]
b.append([3, 4])   # Appends as nested list
print(b)            # [1, 2, [3, 4]]

c = [1, 2]
c.extend([3, 4])   # Extends with individual items
print(c)            # [1, 2, 3, 4]

# extend works with any iterable
d = [1, 2]
d.extend("abc")     # String is iterable → each char added
print(d)            # [1, 2, 'a', 'b', 'c']

# += is shorthand for extend
e = [1, 2]
e += [3, 4]
print(e)  # [1, 2, 3, 4]
```

## 9. Removing Elements — remove, pop, del, clear

Four ways to remove items, each with different behavior.

```python
# --- remove(x) — O(n) ---
# Removes the FIRST occurrence of value x
# Raises ValueError if not found
colors = ["red", "green", "blue", "green"]
colors.remove("green")  # Removes first "green"
print(colors)            # ['red', 'blue', 'green']

try:
    colors.remove("yellow")  # ValueError!
except ValueError as e:
    print(f"ValueError: {e}")
```

```python
# --- pop(index) — O(1) for last, O(n) otherwise ---
# Removes AND RETURNS the item at index (default: last item)
stack = [10, 20, 30, 40]

last = stack.pop()      # Remove last → 40
print(last)             # 40
print(stack)            # [10, 20, 30]

second = stack.pop(1)   # Remove at index 1 → 20
print(second)           # 20
print(stack)            # [10, 30]

# pop() from empty list → IndexError
try:
    [].pop()
except IndexError as e:
    print(f"IndexError: {e}")
```

```python
# --- del statement — O(n) for single, O(n) for slice ---
# delete by index or slice — does NOT return the value
items = ["a", "b", "c", "d", "e"]

del items[0]       # Delete first
print(items)       # ['b', 'c', 'd', 'e']

del items[1:3]     # Delete slice
print(items)       # ['b', 'e']

# del can also delete the entire variable
temp = [1, 2, 3]
del temp
# print(temp)  # NameError — variable no longer exists
```

```python
# --- clear() — O(n) ---
# Removes ALL items, leaving an empty list
data = [1, 2, 3, 4, 5]
data.clear()
print(data)  # []

# Difference: clear() vs reassignment
a = [1, 2, 3]
b = a

a.clear()     # Modifies the object — b sees the change
print(b)      # []

a = [1, 2, 3]
b = a
a = []        # Reassignment — b still points to old object
print(b)      # [1, 2, 3]
```

## 10. Searching — index, count, in

Finding items in a list.

```python
letters = ["a", "b", "c", "b", "d", "b"]

# --- index(x) — returns index of FIRST occurrence ---
print(letters.index("b"))      # 1
print(letters.index("b", 2))   # 3  — start searching from index 2
print(letters.index("b", 4))   # 5  — start searching from index 4

try:
    letters.index("z")  # ValueError!
except ValueError:
    print("'z' not found")

# --- count(x) — returns number of occurrences ---
print(letters.count("b"))  # 3
print(letters.count("z"))  # 0  — no error for missing items

# --- in / not in — boolean membership check ---
print("a" in letters)      # True
print("z" in letters)      # False
print("z" not in letters)  # True
```

```python
# Safe search pattern: check before accessing
target = "x"
if target in letters:
    idx = letters.index(target)
    print(f"Found '{target}' at index {idx}")
else:
    print(f"'{target}' not found")
```

## 11. Sorting — sort vs sorted

This is one of the **most important distinctions** in Python. Know it cold.

| | `list.sort()` | `sorted(iterable)` |
|---|---|---|
| Returns | `None` | New list |
| Mutates? | Yes | No |
| Works on | Lists only | Any iterable |
| Use when | You want to sort in place | You want a sorted copy |

```python
# --- .sort() — in-place, returns None ---
nums = [5, 2, 8, 1, 9]
result = nums.sort()
print(nums)    # [1, 2, 5, 8, 9]
print(result)  # None  <-- WATCH OUT!

# --- sorted() — returns new list ---
nums2 = [5, 2, 8, 1, 9]
result2 = sorted(nums2)
print(nums2)    # [5, 2, 8, 1, 9]  — unchanged
print(result2)  # [1, 2, 5, 8, 9]
```

```python
# Both support reverse and key
words = ["banana", "apple", "cherry", "date"]

# Sort alphabetically
print(sorted(words))  # ['apple', 'banana', 'cherry', 'date']

# Sort in reverse
print(sorted(words, reverse=True))  # ['date', 'cherry', 'banana', 'apple']

# Sort by length
print(sorted(words, key=len))  # ['date', 'apple', 'banana', 'cherry']

# Sort by last character
print(sorted(words, key=lambda w: w[-1]))  # ['banana', 'apple', 'date', 'cherry']

# Case-insensitive sort
mixed = ["Banana", "apple", "Cherry"]
print(sorted(mixed, key=str.lower))  # ['apple', 'Banana', 'Cherry']
```

```python
# sorted() works on ANY iterable
print(sorted("hello"))          # ['e', 'h', 'l', 'l', 'o']
print(sorted({3, 1, 2}))       # [1, 2, 3]
print(sorted(range(5, 0, -1))) # [1, 2, 3, 4, 5]
```

## 12. Reversing — reverse vs reversed vs [::-1]

Three ways to reverse a list, each with different behavior.

```python
# --- .reverse() — in-place, returns None ---
nums = [1, 2, 3, 4, 5]
result = nums.reverse()
print(nums)    # [5, 4, 3, 2, 1]
print(result)  # None

# --- reversed() — returns an iterator (lazy), does NOT mutate ---
nums2 = [1, 2, 3, 4, 5]
rev_iter = reversed(nums2)
print(type(rev_iter))        # <class 'list_reverseiterator'>
print(list(rev_iter))        # [5, 4, 3, 2, 1]
print(nums2)                 # [1, 2, 3, 4, 5]  — unchanged

# --- [::-1] — returns a new reversed list ---
nums3 = [1, 2, 3, 4, 5]
rev_copy = nums3[::-1]
print(rev_copy)  # [5, 4, 3, 2, 1]
print(nums3)     # [1, 2, 3, 4, 5]  — unchanged
```

## 13. Copying — Shallow vs Deep

Understanding copy behavior is crucial for avoiding aliasing bugs.

- **Shallow copy**: Copies the outer list, but inner objects are still references.
- **Deep copy**: Copies everything recursively — fully independent.

```python
# --- Shallow copy methods ---
original = [1, 2, [3, 4]]

copy1 = original[:]        # slice copy
copy2 = original.copy()    # .copy() method
copy3 = list(original)     # constructor copy

# Outer list is independent
copy1.append(5)
print(original)  # [1, 2, [3, 4]]  — no 5

# BUT inner list is SHARED!
copy2[2].append(999)
print(original)  # [1, 2, [3, 4, 999]]  — inner list changed!
print(copy2)     # [1, 2, [3, 4, 999]]
```

```python
import copy

# --- Deep copy — fully independent ---
original = [1, 2, [3, 4]]
deep = copy.deepcopy(original)

deep[2].append(999)
print(original)  # [1, 2, [3, 4]]     — unchanged!
print(deep)      # [1, 2, [3, 4, 999]] — independent
```

## 14. Nested Lists (2D Basics)

Lists can contain other lists, creating multi-dimensional structures.
This is a preview — there's a full module on 2D arrays later.

```python
# A 2D list (matrix / grid)
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Accessing elements: matrix[row][col]
print(matrix[0])       # [1, 2, 3]  — first row
print(matrix[0][1])    # 2          — row 0, col 1
print(matrix[2][2])    # 9          — row 2, col 2

# Iterating over a 2D list
for row in matrix:
    for val in row:
        print(val, end=" ")
    print()  # newline after each row
```

```python
# Get dimensions
rows = len(matrix)
cols = len(matrix[0])  # assumes rectangular
print(f"{rows} x {cols}")  # 3 x 3

# Flatten a 2D list
flat = [val for row in matrix for val in row]
print(flat)  # [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Get a column (e.g., column 1)
col1 = [row[1] for row in matrix]
print(col1)  # [2, 5, 8]
```

## 15. List Comprehensions (Preview)

List comprehensions are one of Python's most powerful features. There's a full module later, but you need the basics now.

```python
# Basic: [expression for item in iterable]
squares = [x ** 2 for x in range(6)]
print(squares)  # [0, 1, 4, 9, 16, 25]

# With condition: [expression for item in iterable if condition]
evens = [x for x in range(10) if x % 2 == 0]
print(evens)  # [0, 2, 4, 6, 8]

# Transform: uppercase all words
words = ["hello", "world"]
upper = [w.upper() for w in words]
print(upper)  # ['HELLO', 'WORLD']

# Nested: flatten
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [x for row in matrix for x in row]
print(flat)  # [1, 2, 3, 4, 5, 6]

# With if/else (transform expression, not filter)
labels = ["even" if x % 2 == 0 else "odd" for x in range(5)]
print(labels)  # ['even', 'odd', 'even', 'odd', 'even']
```

## 16. Unpacking & Multiple Assignment

Unpacking lets you assign list elements to multiple variables at once. This is extremely Pythonic and common in interviews.

```python
# Basic unpacking
coords = [10, 20, 30]
x, y, z = coords
print(x, y, z)  # 10 20 30

# Swap two variables (no temp needed!)
a, b = 1, 2
a, b = b, a
print(a, b)  # 2 1

# Star unpacking — capture the rest
first, *rest = [1, 2, 3, 4, 5]
print(first)  # 1
print(rest)   # [2, 3, 4, 5]

*beginning, last = [1, 2, 3, 4, 5]
print(beginning)  # [1, 2, 3, 4]
print(last)       # 5

first, *middle, last = [1, 2, 3, 4, 5]
print(first)   # 1
print(middle)  # [2, 3, 4]
print(last)    # 5
```

```python
# Unpacking in loops — very common pattern
pairs = [["Alice", 90], ["Bob", 85], ["Charlie", 92]]

for name, score in pairs:
    print(f"{name}: {score}")

# Unpacking with enumerate
fruits = ["apple", "banana", "cherry"]
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
```

```python
# ValueError: mismatch in number of values
try:
    a, b = [1, 2, 3]  # Too many values
except ValueError as e:
    print(f"ValueError: {e}")

try:
    a, b, c = [1, 2]  # Not enough values
except ValueError as e:
    print(f"ValueError: {e}")
```

## 17. Common Built-ins with Lists

These built-in functions work beautifully with lists and show up everywhere.

```python
nums = [4, 7, 2, 9, 1, 5, 8]

# --- len, min, max, sum ---
print(f"len:  {len(nums)}")    # 7
print(f"min:  {min(nums)}")    # 1
print(f"max:  {max(nums)}")    # 9
print(f"sum:  {sum(nums)}")    # 36
print(f"avg:  {sum(nums) / len(nums):.2f}")  # 5.14
```

```python
# --- enumerate — index + value ---
fruits = ["apple", "banana", "cherry"]
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

# Start from a different number
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")
```

```python
# --- zip — iterate multiple lists in parallel ---
names = ["Alice", "Bob", "Charlie"]
scores = [90, 85, 92]

for name, score in zip(names, scores):
    print(f"{name}: {score}")

# zip stops at shortest list
a = [1, 2, 3]
b = ["x", "y"]
print(list(zip(a, b)))  # [(1, 'x'), (2, 'y')]  — 3 is dropped

# Create a dict from two lists
d = dict(zip(names, scores))
print(d)  # {'Alice': 90, 'Bob': 85, 'Charlie': 92}
```

```python
# --- any / all —  boolean checks on list ---
nums = [0, 1, 2, 3]
print(any(nums))   # True  — at least one truthy
print(all(nums))   # False — 0 is falsy

# With conditions
grades = [85, 92, 78, 95, 88]
print(all(g >= 70 for g in grades))  # True — all passing
print(any(g >= 90 for g in grades))  # True — at least one A
```

```python
# --- map / filter — functional style ---
nums = [1, 2, 3, 4, 5]

# map: apply function to each element
doubled = list(map(lambda x: x * 2, nums))
print(doubled)  # [2, 4, 6, 8, 10]

# filter: keep elements that pass the test
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  # [2, 4]

# Usually comprehensions are preferred over map/filter:
doubled2 = [x * 2 for x in nums]
evens2 = [x for x in nums if x % 2 == 0]
```

## 18. Iteration & Traversal Patterns

This is the **most important section for interviews and DSA**. Lists are traversed constantly — know every pattern cold.

### 18.1 Simple Forward Traversal

```python
nums = [10, 20, 30, 40, 50]

# Pattern 1: Iterate by value
for num in nums:
    print(num, end=" ")  # 10 20 30 40 50
print()
```

### 18.2 Index-Based Traversal

```python
# Pattern 2: Iterate by index (when you NEED the index)
for i in range(len(nums)):
    print(f"index {i}: {nums[i]}")
```

### 18.3 Enumerate Traversal

```python
# Pattern 3: enumerate — the Pythonic way to get index + value
for i, num in enumerate(nums):
    print(f"index {i}: {num}")

# This is ALWAYS preferred over range(len(...)) when you need both
```

### 18.4 Reverse Traversal

```python
# Pattern 4a: reversed()
for num in reversed(nums):
    print(num, end=" ")  # 50 40 30 20 10
print()

# Pattern 4b: reverse index range
for i in range(len(nums) - 1, -1, -1):
    print(f"index {i}: {nums[i]}")

# Pattern 4c: slice
for num in nums[::-1]:
    print(num, end=" ")
print()
```

### 18.5 Two-Pointer Traversal

Extremely important for DSA. Two pointers move through the list from different starting points.

```python
# Pattern 5a: Two pointers from both ends (palindrome check)
def is_palindrome(lst):
    left, right = 0, len(lst) - 1
    while left < right:
        if lst[left] != lst[right]:
            return False
        left += 1
        right -= 1
    return True

print(is_palindrome([1, 2, 3, 2, 1]))  # True
print(is_palindrome([1, 2, 3, 4, 5]))  # False
```

```python
# Pattern 5b: Two pointers — same direction (remove duplicates from sorted)
def remove_duplicates(sorted_list):
    if not sorted_list:
        return []
    result = [sorted_list[0]]
    for i in range(1, len(sorted_list)):
        if sorted_list[i] != sorted_list[i - 1]:
            result.append(sorted_list[i])
    return result

print(remove_duplicates([1, 1, 2, 2, 3, 4, 4]))  # [1, 2, 3, 4]
```

### 18.6 Sliding Window Traversal

Process a fixed-size or variable-size window that slides across the list.

```python
# Pattern 6: Fixed-size sliding window (max sum of k consecutive)
def max_subarray_sum(nums, k):
    if len(nums) < k:
        return 0
    
    window_sum = sum(nums[:k])
    max_sum = window_sum
    
    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i - k]  # slide: add right, remove left
        max_sum = max(max_sum, window_sum)
    
    return max_sum

print(max_subarray_sum([2, 1, 5, 1, 3, 2], 3))  # 9 (5+1+3)
```

### 18.7 Neighbor Comparison Traversal

Comparing each element with its neighbor(s). Very common in interview problems.

```python
# Pattern 7a: Compare adjacent pairs
nums = [1, 3, 2, 5, 4]

for i in range(len(nums) - 1):
    if nums[i] > nums[i + 1]:
        print(f"Decrease at index {i}: {nums[i]} > {nums[i+1]}")
# Decrease at index 1: 3 > 2
# Decrease at index 3: 5 > 4
```

```python
# Pattern 7b: Running max/min
nums = [3, 1, 4, 1, 5, 9, 2, 6]
running_max = nums[0]

for i in range(1, len(nums)):
    running_max = max(running_max, nums[i])
    print(f"After index {i}: max so far = {running_max}")
```

### 18.8 Frequency Counting

```python
# Pattern 8a: Manual frequency count with dict
items = ["a", "b", "a", "c", "b", "a"]
freq = {}
for item in items:
    freq[item] = freq.get(item, 0) + 1
print(freq)  # {'a': 3, 'b': 2, 'c': 1}

# Pattern 8b: Counter (preview — there's a full module on this)
from collections import Counter
freq2 = Counter(items)
print(freq2)               # Counter({'a': 3, 'b': 2, 'c': 1})
print(freq2.most_common(2))  # [('a', 3), ('b', 2)]
```

### 18.9 Building a New List from Traversal

```python
# Pattern 9a: Transform each element
nums = [1, 2, 3, 4, 5]
doubled = []
for n in nums:
    doubled.append(n * 2)
print(doubled)  # [2, 4, 6, 8, 10]

# Pattern 9b: Filter elements
words = ["hello", "", "world", "", "foo"]
non_empty = []
for w in words:
    if w:
        non_empty.append(w)
print(non_empty)  # ['hello', 'world', 'foo']

# Comprehension versions (preferred):
doubled2 = [n * 2 for n in nums]
non_empty2 = [w for w in words if w]
```

### 18.10 Multi-List Parallel Traversal

```python
# Pattern 10: zip for parallel iteration
names = ["Alice", "Bob", "Charlie"]
scores = [90, 85, 92]
grades = ["A", "B+", "A"]

for name, score, grade in zip(names, scores, grades):
    print(f"{name}: {score} ({grade})")

# With index
for i, (name, score) in enumerate(zip(names, scores)):
    print(f"{i}. {name} scored {score}")
```

### 18.11 Accumulator Pattern

Build up a result as you traverse.

```python
# Pattern 11a: Sum accumulator
nums = [1, 2, 3, 4, 5]
total = 0
for n in nums:
    total += n
print(total)  # 15

# Pattern 11b: Prefix sum (running total)
prefix = []
running = 0
for n in nums:
    running += n
    prefix.append(running)
print(prefix)  # [1, 3, 6, 10, 15]

# Pattern 11c: String accumulator (join list into string)
chars = ["h", "e", "l", "l", "o"]
result = ""
for c in chars:
    result += c
print(result)  # "hello"

# Better: use join
print("".join(chars))  # "hello"
```

### 18.12 Mutating During Iteration — DANGER!

**Never modify a list while iterating over it.** This is one of the most common bugs.

```python
# ❌ BAD: Removing while iterating forward
nums = [1, 2, 3, 4, 5, 6]
for n in nums:
    if n % 2 == 0:
        nums.remove(n)
print(nums)  # [1, 3, 5, 6] — WRONG! Missed 6!

# Why? When you remove 2, everything shifts left.
# The iterator skips 3 and jumps to 4.
# When you remove 4, it skips 5 and 6.
```

```python
# ✅ FIX 1: Iterate over a copy
nums = [1, 2, 3, 4, 5, 6]
for n in nums[:]:  # slice copy
    if n % 2 == 0:
        nums.remove(n)
print(nums)  # [1, 3, 5]

# ✅ FIX 2: Build a new list with comprehension (preferred)
nums = [1, 2, 3, 4, 5, 6]
odds = [n for n in nums if n % 2 != 0]
print(odds)  # [1, 3, 5]

# ✅ FIX 3: Iterate backwards (safe for removal by index)
nums = [1, 2, 3, 4, 5, 6]
for i in range(len(nums) - 1, -1, -1):
    if nums[i] % 2 == 0:
        nums.pop(i)
print(nums)  # [1, 3, 5]
```

## 19. Common Gotchas & Pitfalls

Master these to avoid bugs that plague beginners and intermediates alike.

```python
# Gotcha 1: Mutable default arguments
def add_item(item, lst=[]):  # ❌ BAD — the default list is SHARED
    lst.append(item)
    return lst

print(add_item(1))  # [1]
print(add_item(2))  # [1, 2]  — BUG! Expected [2]
print(add_item(3))  # [1, 2, 3]  — BUG! The same list persists

# ✅ FIX: Use None as default
def add_item_fixed(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(add_item_fixed(1))  # [1]
print(add_item_fixed(2))  # [2]  ✓
```

```python
# Gotcha 2: sort() returns None
nums = [3, 1, 2]
# ❌ BAD
result = nums.sort()
print(result)  # None

# ✅ FIX: Use sorted() if you need the return value
nums = [3, 1, 2]
result = sorted(nums)
print(result)  # [1, 2, 3]
```

```python
# Gotcha 3: Comparing lists — element-wise, not length
print([1, 2, 3] == [1, 2, 3])   # True
print([1, 2, 3] == [3, 2, 1])   # False (order matters)
print([1, 2] < [1, 3])          # True (compares element-by-element)
print([1, 2] < [1, 2, 0])       # True (shorter list is "less")
```

```python
# Gotcha 4: Empty list is falsy
lst = []
if not lst:
    print("Empty!")  # This prints

# Prefer this over len(lst) == 0
# Pythonic: `if lst:` (truthy) or `if not lst:` (falsy)
```

```python
# Gotcha 5: Off-by-one errors with range
nums = [10, 20, 30, 40, 50]

# ❌ BAD: range(len(nums) - 1) misses the last element
for i in range(len(nums) - 1):  # 0, 1, 2, 3 — misses index 4!
    pass

# ✅ Correct: range(len(nums)) goes 0 to 4
for i in range(len(nums)):  # 0, 1, 2, 3, 4
    pass

# NOTE: range(len(nums) - 1) IS correct when comparing neighbors (nums[i] vs nums[i+1])
```

```python
# Gotcha 6: += with lists vs =
a = [1, 2]
b = a

a += [3, 4]    # Mutates a IN PLACE (like extend)
print(b)        # [1, 2, 3, 4]  — b changed!

a = [1, 2]
b = a

a = a + [3, 4]  # Creates a NEW list, reassigns a
print(b)         # [1, 2]  — b unchanged
```

## 20. Lists vs Strings — Comparison

| Feature | Strings | Lists |
|---------|---------|-------|
| Mutable? | ❌ No | ✅ Yes |
| Indexing | `s[0]` | `lst[0]` |
| Slicing | `s[1:3]` | `lst[1:3]` |
| `in` operator | `"he" in "hello"` (substring) | `1 in [1,2,3]` (exact element) |
| Iteration | char by char | element by element |
| Concatenation | `"a" + "b"` → new string | `[1] + [2]` → new list |
| Convert to other | `list("abc")` → `['a','b','c']` | `"".join(['a','b','c'])` → `"abc"` |

```python
# String ↔ List conversion
s = "hello world"

# String → List of chars
chars = list(s)
print(chars)  # ['h', 'e', 'l', 'l', 'o', ' ', 'w', 'o', 'r', 'l', 'd']

# String → List of words
words = s.split()
print(words)  # ['hello', 'world']

# List of chars → String
back = "".join(chars)
print(back)  # "hello world"

# List of words → String
sentence = " ".join(words)
print(sentence)  # "hello world"

# Common pattern: modify a "string" via list
s = "hello"
chars = list(s)
chars[0] = "H"
s = "".join(chars)
print(s)  # "Hello"
```

## 21. Real-World Use Cases

Lists are everywhere in production code.

```python
# Backend: Processing API request data
raw_tags = [" Python ", "  ", "FastAPI", "", " Data Science "]
clean_tags = [tag.strip() for tag in raw_tags if tag.strip()]
print(clean_tags)  # ['Python', 'FastAPI', 'Data Science']

# Data: Filtering records
users = [
    {"name": "Alice", "active": True},
    {"name": "Bob", "active": False},
    {"name": "Charlie", "active": True},
]
active_users = [u["name"] for u in users if u["active"]]
print(active_users)  # ['Alice', 'Charlie']

# Interviews: Two Sum (classic)
def two_sum(nums, target):
    seen = {}  # value → index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

print(two_sum([2, 7, 11, 15], 9))  # [0, 1]
```

```python
# ETL: Batch processing
raw_data = ["100", "abc", "200", "xyz", "300"]

cleaned = []
errors = []
for item in raw_data:
    if item.isdigit():
        cleaned.append(int(item))
    else:
        errors.append(item)

print(f"Cleaned: {cleaned}")  # [100, 200, 300]
print(f"Errors: {errors}")    # ['abc', 'xyz']

# Queue simulation (preview — deque is better)
task_queue = ["job_a", "job_b", "job_c"]
while task_queue:
    current = task_queue.pop(0)  # O(n) — deque is O(1)
    print(f"Processing: {current}")
```

## 22. Summary

### Key Takeaways

1. **Lists are mutable** — strings are not. This changes everything.
2. **Methods that mutate return None** — `sort()`, `reverse()`, `append()`, etc.
3. **Aliasing is dangerous** — `b = a` doesn't copy, it creates a reference.
4. **Know the Big O** — `append()` is O(1), `insert(0, x)` is O(n), `in` is O(n).
5. **Never mutate during forward iteration** — use copy, comprehension, or reverse iteration.
6. **Mutable default arguments are bugs waiting to happen** — use `None`.
7. **`sorted()` vs `.sort()`** — know which returns a value and which doesn't.
8. **Traversal patterns are the foundation of DSA** — master all 12 patterns above.
9. **List comprehensions** are the Pythonic replacement for transform/filter loops.
10. **Deep copy vs shallow copy** matters with nested lists.
