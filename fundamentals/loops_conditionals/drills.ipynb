```python
# Explore 5: Nested ternary vs if/elif/else
# Compare these two approaches:

score = 75

# Approach 1: if/elif/else
if score >= 90:
    grade1 = "A"
elif score >= 80:
    grade1 = "B"
elif score >= 70:
    grade1 = "C"
else:
    grade1 = "F"

# Approach 2: Nested ternary
grade2 = "A" if score >= 90 else "B" if score >= 80 else "C" if score >= 70 else "F"

print(f"if/elif: {grade1}")
print(f"ternary: {grade2}")
print(f"Match: {grade1 == grade2}")

# Question: Which is more readable? When would you use each?

# Your explanation:
```

```python
# Explore 6: range() is lazy — it doesn't create a list in memory
import sys

r = range(1_000_000)
l = list(range(1_000_000))

print(f"range object size: {sys.getsizeof(r)} bytes")
print(f"list size:         {sys.getsizeof(l)} bytes")

# You can also check membership in range efficiently
print(f"999999 in range(1_000_000): {999999 in r}")  # O(1)!
print(f"999999 in list:             {999999 in l}")    # O(n)

# Question: Why is the range object so much smaller? 
# Why is 'in' check O(1) for range but O(n) for list?

# Your explanation:
```
