```python
# Debug 1: Two Sum — returns wrong answer for [3, 3], target=6
# BUG: How does it handle duplicate values?

def two_sum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        seen[num] = i              # Store BEFORE checking
        complement = target - num
        if complement in seen and seen[complement] != i:
            return [seen[complement], i]
    return []

# Test: should be [0, 1] but what happens?
print(two_sum([3, 3], 6))

# What's the bug? How do you fix the order of operations?
```

```python
# Debug 7: Anagram check gives wrong result for different lengths
# BUG: Says "abc" and "abcd" are NOT anagrams (correct), but WHY
# does this approach fail without the length check?

def is_anagram(s, t):
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    for ch in t:
        freq[ch] = freq.get(ch, 0) - 1
    return all(v == 0 for v in freq.values())

print(is_anagram("abc", "cba"))    # True ✓
print(is_anagram("abc", "abcd"))   # Should be False
print(is_anagram("abcd", "abc"))   # Should be False — does it work?

# This version actually works! But what if someone adds an early return?
# When would you NEED the length check?
```
