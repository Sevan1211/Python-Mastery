## Section 2: Identify the Pattern

For each problem description, identify which Two Pointer sub-pattern to use and explain WHY.

```python
# Pattern ID 1:
# "Given a sorted array, find two numbers that add up to a target."
#
# Your answer: Which pattern? Why?
# 
# Answer: Opposite Ends. Array is sorted, looking for a pair with sum condition.
# Place pointers at both ends. Adjust based on current sum vs target.
```

```python
# Pattern ID 2:
# "Given a sorted array, remove duplicates in-place and return the new length."
#
# Your answer: Which pattern? Why?
#
# Answer: Slow/Fast. In-place removal + sorted array.
# Slow marks write position, fast scans for new unique values.
```

```python
# Pattern ID 3:
# "Given an array of 0s, 1s, and 2s, sort them in a single pass."
#
# Your answer: Which pattern? Why?
#
# Answer: Dutch National Flag. Three categories to partition in-place.
# Use three pointers: low, mid, high. Single pass.
```

```python
# Pattern ID 4:
# "Find all unique triplets in an array that sum to zero."
#
# Your answer: Which pattern? Why?
#
# Answer: Multi-Sum (3Sum). Fix one element, use two pointers on the rest.
# Sort first, skip duplicates at all levels.
```

```python
# Pattern ID 5:
# "Merge two sorted arrays into one sorted array."
#
# Your answer: Which pattern? Why?
#
# Answer: Merge Two Sorted. One pointer per array, pick the smaller element.
# Don't forget the remainder after one pointer finishes.
```

```python
# Pattern ID 6:
# "Check if a string is a palindrome, ignoring non-alphanumeric characters."
#
# Your answer: Which pattern? Why?
#
# Answer: String Two Pointers. Compare characters from both ends, moving inward.
# Skip non-alphanumeric chars with inner while loops.
```

```python
# Pattern ID 7:
# "Given a string s and a string t, check if s is a subsequence of t."
#
# Your answer: Which pattern? Why?
#
# Answer: Same Direction (one pointer per sequence). Walk through both strings,
# advance the s pointer only on matches, always advance the t pointer.
```

```python
# Pattern ID 8:
# "Given an array of heights representing bars, calculate how much rainwater
#  can be trapped."
#
# Your answer: Which pattern? Why?
#
# Answer: Opposite Ends with max tracking. Process the shorter side first.
# Water at each position depends on min(left_max, right_max) - height.
# By processing the shorter side, we know the bottleneck.
```

## Section 3: Debug the Bug

Each cell has a bug. Find it, explain it, then fix it.

```python
# Debug 1: Two Sum II on sorted array
def two_sum_sorted(numbers, target):
    left, right = 0, len(numbers)  # BUG: should be len(numbers) - 1
    while left < right:
        s = numbers[left] + numbers[right]
        if s == target:
            return [left, right]
        elif s < target:
            left += 1
        else:
            right -= 1
    return []

# Fix: change len(numbers) to len(numbers) - 1
# Why: right pointer starts out of bounds, causing IndexError

# Uncomment to test the bug:
# two_sum_sorted([2, 7, 11, 15], 9)
```

```python
# Debug 2: Remove duplicates
def remove_duplicates(nums):
    if not nums:
        return 0
    slow = 0
    for fast in range(len(nums)):  # BUG: should start at 1
        if nums[fast] != nums[slow]:
            slow += 1
            nums[slow] = nums[fast]
    return slow + 1

# Test: this won't crash but fast=0 compares nums[0] with nums[0]
# which always matches — so it wastes a comparison but still works.
# The real bug: it's not wrong here, but starting at 0 is wasteful.
# Fixed: range(1, len(nums))

arr = [1, 1, 2]
print(remove_duplicates(arr), arr)  # Works but inefficient
```

```python
# Debug 3: 3Sum missing duplicate skip
def three_sum_buggy(nums):
    nums.sort()
    result = []
    for i in range(len(nums) - 2):
        # BUG: No duplicate check for i!
        left, right = i + 1, len(nums) - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total == 0:
                result.append([nums[i], nums[left], nums[right]])
                left += 1
                right -= 1
            elif total < 0:
                left += 1
            else:
                right -= 1
    return result

# Test: produces duplicates!
print(three_sum_buggy([-1, 0, 1, 2, -1, -4]))
# Gets [[-1, -1, 2], [-1, 0, 1], [-1, 0, 1]] — duplicate!

# Fix: Add at the start of the loop:
#   if i > 0 and nums[i] == nums[i-1]: continue
# Also add left/right duplicate skipping after finding a match.
```

```python
# Debug 4: Dutch Flag — incrementing mid when swapping with high
def sort_colors_buggy(nums):
    low, mid, high = 0, 0, len(nums) - 1
    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1
            mid += 1  # BUG: should NOT increment mid here!
    return nums

# Test:
print(sort_colors_buggy([2, 0, 1]))  # Might work by luck
print(sort_colors_buggy([2, 2, 0, 0, 1]))  # Will fail!

# Fix: Remove 'mid += 1' from the else branch.
# Why: The element swapped from high is unexplored — we must check it again.
```

```python
# Debug 5: Move zeroes — overwrite instead of swap
def move_zeroes_buggy(nums):
    slow = 0
    for fast in range(len(nums)):
        if nums[fast] != 0:
            nums[slow] = nums[fast]  # BUG: this overwrites, doesn't swap
            slow += 1
    return nums

# Test:
print(move_zeroes_buggy([0, 1, 0, 3, 12]))
# Gets [1, 3, 12, 3, 12] — non-zeros are copied but old values remain!

# Fix: Either swap (nums[slow], nums[fast] = nums[fast], nums[slow])
# Or do a second pass filling zeros: nums[slow:] = [0] * (len(nums) - slow)
```

```python
# Debug 6: Merge sorted array — wrong direction
def merge_buggy(nums1, m, nums2, n):
    p1, p2 = 0, 0  # BUG: merging from the front overwrites!
    write = 0
    temp = nums1[:m]  # Need a copy to avoid overwriting
    while p1 < m and p2 < n:
        if temp[p1] <= nums2[p2]:
            nums1[write] = temp[p1]
            p1 += 1
        else:
            nums1[write] = nums2[p2]
            p2 += 1
        write += 1
    while p1 < m:
        nums1[write] = temp[p1]
        p1 += 1
        write += 1
    while p2 < n:
        nums1[write] = nums2[p2]
        p2 += 1
        write += 1
    return nums1

# This works but uses O(m) extra space. The O(1) approach is to merge from the BACK.
# Fix: Start p1=m-1, p2=n-1, write=m+n-1, compare largest first.

print(merge_buggy([1, 2, 3, 0, 0, 0], 3, [2, 5, 6], 3))
```

```python
# Debug 7: Palindrome — wrong loop condition
def is_palindrome_buggy(s):
    left, right = 0, len(s) - 1
    while left <= right:  # BUG: should be left < right
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

# For odd-length strings, when left == right, you compare s[mid] with itself.
# That's not really a bug — it works correctly but does one unnecessary comparison.
# For even-length strings, left > right after they cross, so <= causes one extra
# iteration where they've already crossed.

# This is a subtle "not quite a bug" — it works but is slightly wrong conceptually.
# Fix: use left < right (they should stop before meeting).

print(is_palindrome_buggy("racecar"))  # True (still works)
print(is_palindrome_buggy("ab"))        # False (still works)
```
