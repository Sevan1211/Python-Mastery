# Array Traversal — Drills

These drills hammer the **mechanics** of looping over arrays. They are NOT full interview problems — they're micro-exercises that build muscle memory for indexing, state-tracking, neighbour comparison, in-place writes, and direction-aware scans.

Five sections:
1. **Predict the output** — read the loop, predict before running.
2. **Fill in the blank** — complete a loop body or update step.
3. **Debug the bug** — find the off-by-one or mutation bug.
4. **Build from scratch** — write the loop yourself.
5. **Explore** — vary inputs and observe.

Run each cell, predict the answer FIRST, then check.

## Section 3 — Debug the Bug

Each cell has a subtle bug. Find it and fix in place. Then run.

```
# D3.1 — sum is wrong on every input
def my_sum(arr):
    total = 0
    for x in arr:
        total = x       # BUG
    return total

# Expected: my_sum([1,2,3]) == 6
print(my_sum([1, 2, 3]))   # currently prints 3
```

```
# D3.2 — argmax returns the value, not the index
def argmax(arr):
    best_i = 0
    for i in range(len(arr)):
        if arr[i] > arr[best_i]:
            best_i = arr[i]    # BUG: should be i
    return best_i

# Expected: argmax([3,1,4,1,5,9,2,6]) == 5
print(argmax([3, 1, 4, 1, 5, 9, 2, 6]))
```

```
# D3.3 — off by one: out-of-bounds on the last iteration
def neighbour_diffs(arr):
    out = []
    for i in range(len(arr)):           # BUG: should stop one earlier
        out.append(arr[i+1] - arr[i])
    return out

# Expected: [10, 10, 10] for [1,11,21,31]
print(neighbour_diffs([1, 11, 21, 31]))
```

```
# D3.4 — mutating list while iterating: skips elements
def remove_zeros(arr):
    for x in arr:
        if x == 0:
            arr.remove(x)        # BUG: modifying during iteration
    return arr

# Expected: [1, 2, 3]
print(remove_zeros([0, 1, 0, 0, 2, 3]))
```

```
# D3.5 — Kadane forgets to reset
def max_subarray(arr):
    cur = best = arr[0]
    for x in arr[1:]:
        cur = cur + x        # BUG: never resets when negative
        best = max(best, cur)
    return best

# Expected: 6 for [-2,1,-3,4,-1,2,1,-5,4]
print(max_subarray([-2, 1, -3, 4, -1, 2, 1, -5, 4]))
```

```
# D3.6 — best_profit fails on monotonically decreasing prices
def max_profit(prices):
    min_price = prices[0]
    best = prices[0]              # BUG: initial best should be 0
    for p in prices[1:]:
        best = max(best, p - min_price)
        min_price = min(min_price, p)
    return best

# Expected: 0 for [7,6,4,3,1]
print(max_profit([7, 6, 4, 3, 1]))
```

```
# D3.7 — max_consecutive_ones doesn't reset on 0
def mco(nums):
    best = run = 0
    for x in nums:
        if x == 1:
            run += 1             # BUG: missing else branch resets
        best = max(best, run)
    return best

# Expected: 3 for [1,1,0,1,1,1,0,1]  (with bug it returns 6)
print(mco([1, 1, 0, 1, 1, 1, 0, 1]))
```

```
# D3.8 — empty input crashes
def find_min(arr):
    best = arr[0]              # BUG: crashes on []
    for x in arr[1:]:
        if x < best:
            best = x
    return best

# Expected: None on empty input
try:
    print(find_min([]))
except Exception as e:
    print('crashed:', e)
```

## Section 5 — Explore

Open-ended cells. Vary inputs and observe.

```
# D5.1 — what's faster: a manual loop or sum()?  (mostly observational)
import timeit
arr = list(range(100_000))
t1 = timeit.timeit(lambda: sum(arr), number=200)
t2 = timeit.timeit(lambda: sum(x for x in arr), number=200)
def manual():
    s = 0
    for x in arr:
        s += x
    return s
t3 = timeit.timeit(manual, number=200)
print(f"sum(arr)        : {t1:.4f}")
print(f"sum(x for x in) : {t2:.4f}")
print(f"manual loop     : {t3:.4f}")
```

```
# D5.2 — enumerate vs range(len(arr)) — both work, but enumerate is idiomatic
arr = ['a', 'b', 'c']
for i, x in enumerate(arr):
    print(i, x)
print('---')
for i in range(len(arr)):
    print(i, arr[i])
```

```
# D5.3 — copy vs reference; mutating a slice doesn't mutate the original
a = [1, 2, 3, 4, 5]
b = a[:]      # full slice = shallow copy
b[0] = 99
print('a =', a)
print('b =', b)
```

```
# D5.4 — modifying a list while iterating: surprising behavior
a = [1, 2, 3, 4, 5]
for x in a:
    if x % 2 == 0:
        a.remove(x)
print(a)   # NOT [1, 3, 5] — try predicting before running
```

```
# D5.5 — reversed() is a generator, not a list — print the type
r = reversed([1, 2, 3])
print(type(r))
print(list(r))
```
