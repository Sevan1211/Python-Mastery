# Array Traversal — Practice Problems

**30 problems across 5 levels.** Each problem includes:

- A function stub for you to complete
- Inline assert tests + a pass message
- A header comment showing the **brute force**, the **optimal complexity target**, and the **edge cases** to consider
- A "production analogy" comment so you remember where this pattern actually shows up in real systems

### Workflow per problem

1. Read the problem and the brute-force note.
2. List edge cases out loud (empty, single, all same, negatives, sorted/unsorted).
3. State the brute force *and* its complexity, then the optimized target.
4. Write the optimal solution.
5. Run the cell. If it fails, debug — don't peek at the lesson unless stuck for >10 min.

### Difficulty levels

- **Level 1 (Problems 1–6):** Warm-up traversal mechanics.
- **Level 2 (Problems 7–13):** Single-pass aggregation & best-so-far.
- **Level 3 (Problems 14–20):** In-place modification & two-pointer-style traversals.
- **Level 4 (Problems 21–26):** Classic interview problems (Kadane, Pivot, Product Except Self, etc.).
- **Level 5 (Problems 27–30):** Harder / multi-pass / direction-aware.
